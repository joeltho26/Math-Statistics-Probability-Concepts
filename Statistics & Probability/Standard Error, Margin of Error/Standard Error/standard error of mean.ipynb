#### Standard Error of Mean:
- used to compare between the mean value of sample and population and identify the amount of error

##### if the mean & standard deviation of population is known then,

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/standard-error2.png')
new_image = img.resize((220, 120))
display(new_image)
```

Example: Math SAT Score
- n = 200 students (sample from population) 
- μ = 550 marks (population)
- σ = 180 marks (population)
- SEM = 12.8

Soln:
- if we take many different samples of 200 students
- the average maths score of those sample means likely vary about 12.8 marks from the true population mean of math score.
    i.e. 550 +- 12.8

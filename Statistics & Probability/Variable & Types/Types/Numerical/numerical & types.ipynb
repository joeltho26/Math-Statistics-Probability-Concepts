#### Numerical Variable Types:
- ```Numerical```
    - has a numerical representation
    - has 2 sub-types
        - ```Discrete```
            - the values are distinct
            - eg: outcomes of a coin flip, number of students in a class room, age of a person, etc
        - ```Continuous```
            - the values are not restricted and can also contain decimal values too
            - eg: height/weight of a person

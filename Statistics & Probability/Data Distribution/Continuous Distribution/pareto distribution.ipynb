#### Pareto Distribution:
- it also known as 80/20 rule
- 20% change in 1 variable constitutes 80% change in another variable or 20% of variable 1 constitute to 80% of the other variable
- converted to normal distribution via box cox transformation
- α is the hyper parameter
    - higher α = steeper curve and eventually straight line
- eg:
    - 20% of customers contribute to 80% of the sales
- it starts from a higher value and then slopes down to a lower value

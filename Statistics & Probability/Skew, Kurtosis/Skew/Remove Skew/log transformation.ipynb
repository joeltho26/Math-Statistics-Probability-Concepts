#### Log Transformation - Gaussian:
- reduces skewness and kurtosis of data
- log base e is applied to the data for transformation
- only works with positive values and if negative and zero values are applied to log it results in undefined error
- helps in outlier detection

#### Handle Zero & Negative Values for Log Transformation:
- Absolute Values: convert the negative value as absolute values by only choosing the magnitude, but will result losing information by negating the sign.
    - eg: -3.5 => 3.5
- Adding Constant: add a constant value to all the values to convert the negative and zero value to positive values
- Reflect Values: multiple the negative values with -1 to produce positive values

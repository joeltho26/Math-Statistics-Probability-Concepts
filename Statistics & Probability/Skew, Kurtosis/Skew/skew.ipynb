#### Skew:
- where the distribution of a variable in a dataset can be either extended in left or right tail
- 2 types of skewness
    - positive/right
    - negative/left
- it also sometimes indicate outliers
- it is necessary to address the skewness of a data

#### Positive Skew:
- where the distribution is extended to the right tail
- also known as right skewed
- mode > median > mean
- eg: income distribution, return of investments, etc

#### Negative Skew:
- where the distribution is extended to the left tail
- also known as left skewed
- mode > median > mean
- eg: no. of deaths distribution, Pregnancy months, retirement age, etc

#### No Skew:
- we can also have data distribution with no skewness in it
- distribution is symmetric on both sides from the mean value
- mean = median = mode

#### Histogram:
- it is used to plot both numerical and categorical values
- the histogram contains the details about the frequency of each values in a variable
    - numerical:
        - continuous values: creates artificial bins to hold a certain range of values
        - discrete values: create a separate bar for each value
    - categorical:
        - nominal or ordinal: create a separate bar for each value
- smoothening the histogram will create a density plot (just like a normal distribution of data)

#### Karl Pearson Coefficient:
- used to measure the strength and direction of skewness 
- calculated using both mode or median
- used for discrete numerical data and categorical data

#### First Pearson coefficent of skewness:
- using mode for first coefficient for skewness
- suitable when a strong mode is exhibited by the sample data

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/mode_pearson.png')
new_image = img.resize((340, 90))
display(new_image)
```

#### Second Pearson coefficent of skewness
- using median is called second coefficient of skewness
- suitable when a multiple mode or weak mode is exhibited by the sample data

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/median_pearson.png')
new_image = img.resize((340, 90))
display(new_image)
```

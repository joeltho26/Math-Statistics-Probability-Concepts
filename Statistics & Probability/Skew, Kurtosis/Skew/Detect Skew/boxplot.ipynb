#### Boxplot:
- a plot used to identify skewness of data distribution for a variable
- plot implemented for IQR

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/boxplot.png')
new_image = img.resize((440, 280))
display(new_image)
```

Inside the Boxplot:
- the dotted line represents the mean
- the solid represents the median 

Outside the Boxplot:
- the two whiskers outside the box plot on either side measures the maximum and minimum values of a variable distribution

Outside the whiskers:
- often they are considered are outliers

Eg: 
- Aisle: right/positive skewed
- Bathroom: slightly left/negative skewed
- Recreation room: right/positive skewed
- Hospital room: almost symmetrical

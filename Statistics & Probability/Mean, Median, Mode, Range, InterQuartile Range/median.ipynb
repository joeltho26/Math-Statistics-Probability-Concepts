#### Median:
- calculates the middle value of the dataset
- it is not impacted by outliers
- median has 2 types of evaluation 
    - Odd values:
        - eg: n = 1,2,3,4,5
            - the middle value is 3
    - Even Values:
        - eg: n = 1,2,3,4,5,6
            - the middle value is 5

#### Geometric Median:
- minimizes the sum of distances to all the data points
- it is used in 2 or more dimensional space (Euclidean space)

#### Weighted Median:
- there is a weight given to the each data point in the dataset
- the weight value can be associated with the frequency of the value, etc
- specific importance is given to certain values in the dataset

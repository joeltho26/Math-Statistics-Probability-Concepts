#### Mode:
- measures the frequency of the values in a variable and return the value with the highest frequency

#### Unimodal:
- variable that contains one value with the highest frequency

#### Bimodal: 
- variable that contains two values with the highest frequency
- understand data distribution and also distinguish distinct groups

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/bimodalqq.png')
new_image = img.resize((340, 250))
display(new_image)
```

#### Trimodal:
- variable that contains three values with the highest frequency
- understand data distribution and also distinguish distinct groups

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/mode.png')
new_image = img.resize((820, 320))
display(new_image)
```

#### Multimodal:
- a variable that consist of more than 3 highest frequency of values
- understand data distribution and also distinguish distinct groups
- eg: 13, 34, 12, 55, 8, 10, 13, 34, 55, 4, 8, 13, 68, 34, 55, 8, 13, 74, 34, 55, 8

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/multi.png')
new_image = img.resize((320, 220))
display(new_image)
```

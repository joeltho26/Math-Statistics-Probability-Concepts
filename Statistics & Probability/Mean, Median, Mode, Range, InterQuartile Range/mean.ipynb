#### Mean:
- denoted as the ratio of sum of numbers or quantities by the total number of quantities or numbers 
- measure the average of all the data points or number or quantities
- it is impacted by outliers
- mean is used in several statisical calculation such as distances, dispersion of data, etc
- mostly used with numerical continuous values

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/mean.png')
new_image = img.resize((220, 120))
display(new_image)
```

#### Geometric Mean:
- denoted as the nth root of product of all the numbers or quantities
    - nth value is denoted by the total number of quantities or numbers
- it is applicable when dealing with values related to growth, rate of change, compounding or have mulitiplicative factor

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/gmean.png')
new_image = img.resize((390, 120))
display(new_image)
```

#### Harmonic Mean:
- it is denoted as the ratio of total count of number or quantities by the reciprocal sum of individual number or quantities
- it is used when dealing with profit loss, speed, etc

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/hmean.png')
new_image = img.resize((350, 120))
display(new_image)
```

#### Weighted Mean:
- it is that there is a weight value associated with each quantities or values
- denoted by the ratio of sum of product of weight and quantities or numbers by the sum of total weight
- it is used when we want apply specific importance to certain quantities and numbers based on frequency, etc

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/wmean.png')
new_image = img.resize((320, 120))
display(new_image)
```

#### Relation of GM, AM, HM:
- HM < GM < AM
- HM * AM = GM^2

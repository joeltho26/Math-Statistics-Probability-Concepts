#### InterQuartile Range:
- used to detect outliers, skewness, kurtosis of data
- consist of 25th quantile (Q1), 50th quantile (Q2), 75th quantile (Q3)
- IQR = Q3 - Q1
- consist of min, max value of the variable to identify the extreme values i.e. outliers denoted by whiskers
    - min value = Q1 - 1.5 * IQR
    - max value = Q3 + 1.5 * IQR
- similar to Standard deviation, Variance, it is also used to measure the variablitiy of the values within a variable in a dataset
- mostly used with numerical continuous values

#### Process of InterQuartile Range:

Values for a variable in a dataset 
- eg: 15, 38, 20, 19, 31, 41, 25

Steps:
1. sort the data and calculate the median of the data values (Q2 i.e. 50th percentile)
    - 15, 19, 20, 25, 31, 38, 41
    - median: since odd number of sample values
        - (n+1)/2 => 25 
2. the value of Q2 is 25 as calculated above
3. now calculate Q1 (25th percentile) & Q2 (75th percentile)
    - calculate Q1:
        - conside the numbers from beginning to numbers less than Q2
        - Repeat step 1
        - Q1 => 19
    - calculate Q2:
        - conside the numbers from after Q2 and till the end
        - Repeat step 1
        - Q3 => 38
4. we have calculated the value of Q1 = 19, Q2 = 25, Q3 = 38 
5. Calculate the IQR value
    - IQR = Q3-Q1 => 19
6. calculate the min and max value (whiskers)
    - min value = -9.5
    - max value = 66.5

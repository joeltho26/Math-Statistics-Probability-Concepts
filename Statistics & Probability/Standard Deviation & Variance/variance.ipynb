#### Variance:
- it is used to measure dispersion of data or how far the data point are from its mean value
- denoted by the average sum of squared difference between the data points and mean value
- it is the square of standard deviation
- it is measured in squared units when compared to original data unit 
- application of variance is more convenient in mathematical analysis

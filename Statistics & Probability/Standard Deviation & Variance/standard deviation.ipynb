#### Standard Deviation:
- it is used to measure the dispersion for data or how far are the data point from its mean value
- denoted by the square root of sum of average distances between the data points and the mean value 
- is the square root of variance
- used to identify outlier
- used for distribution of data (via pdf)
- helps to understand the data intutively
- uses the same units of measurement for evaluation as the the original units of data
- used for numerical continuous values

#### Chi-squared test of independence:
- used to compare 2 categorical values i.e. 2 categorical variables with 2 groups
- used to measure the proportion of both the categories and identify them if they are statistically different
- to check if there is any association between the variables

##### Example:
Question:
- Categorical variable A
    - Non Smoker
    - Smoker

- Categorical Variable B
    - Regular Exercise
    - No Regular Exercise

Answer:
- Null Hypothesis: there is no relationship between the variables
- Alternate Hypothesis: there is relationship with variables
- calculate the expected value
    - Expected Value (Regular Exercise, Smoker) = 35
    - Expected Value (Regular Exercise, Non-Smoker) = 65
    - Expected Value (No Regular Exercise, Smoker) = 35
    - Expected Value (No Regular Exercise, Non-Smoker) = 65

|                     | Smoker | Non-Smoker | Total |
|---------------------|----------------|------------|-------|
| Regular Exercise | 35             | 65         | 100   |
| No Regular Exercise | 35             | 65         | 100   |
| Total | 70             | 130        | 200   |
- Calculate the Chi-Square Test Statistic (χ²):
    - ((25 - 35)² / 35) = (-10)² / 35 = 100 / 35 ≈ 2.857
    - ((75 - 65)² / 65) = (10)² / 65 = 100 / 65 ≈ 1.538
    - ((45 - 35)² / 35) = (10)² / 35 = 100 / 35 ≈ 2.857
    - ((55 - 65)² / 65) = (-10)² / 65 = 100 / 65 ≈ 1.538
    - χ² ≈ 2.857 + 1.538 + 2.857 + 1.538 ≈ 8.79
- calculate the Degrees of Freedom (df)
    df = (n1-1) * (n2-1) => 1
- calculate the p-value
    - using the degree of freedom (1) and test statistic (8.79)
    - p-value = 0.003
- Thus P-value(0.003) < significance level (0.05), we reject the null hypothesis.

#### One-way Anova Test:
- used to compare the means of 1 categorical variable with more than 2 groups or taking more than 2 samples and determine if those from the same population distribution

##### Example:
Question:
- sample size = n1 = n2 = n3 = 5
- 1 categorical variable with 3 groups:
    - Fertilizer A
    - Fertilizer B
    - Fertilizer C

Answer:
- Null Hypothesis: Fertilizer A, B, C have same mean 
- Alternate Hypothesis: Fertilizer A, B, C do not have same mean
- significance level = 0.05

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/one_way_anova_data.png')
new_image = img.resize((430, 250))
display(new_image)
```

- calculate the mean of each sample and also the overall mean by combining all the groups
    - Mean for Fertilizer A = (15 + 17 + 16 + 14 + 18) / 5 = 16 cm
    - Mean for Fertilizer B = (18 + 20 + 19 + 21 + 22) / 5 = 20 cm
    - Mean for Fertilizer C = (12 + 14 + 13 + 11 + 15) / 5 = 13 cm
    - Grand Mean = (15+17+16+14+18+18+20+19+21+22+12+14+13+11+15) / 15 = 16.33 cm
- calculate the sum of squares
    - calculate sum of squares Between Groups
        - k = number of groups (3)
        - (Nj) = number of observations in group (j) (5 for each group)
        - (Xj) = mean of group
        - (Xg) = overall mean
    - SSB= 5(16−16.33)2 + 5(20−16.33)2 + 5(13−16.33)2 = 123.33

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/SSB.png')
new_image = img.resize((250, 100))
display(new_image)
```

- calculate sum of squares Within Groups
    - For Fertilizer A: ((15-16)^2 + (17-16)^2 + (16-16)^2 + (14-16)^2 + (18-16)^2 = 1 + 1 + 0 + 4 + 4 = 10)
    - For Fertilizer B: ((18-20)^2 + (20-20)^2 + (19-20)^2 + (21-20)^2 + (22-20)^2 = 4 + 0 + 1 + 1 + 4 = 10)
    - For Fertilizer C: ((12-13)^2 + (14-13)^2 + (13-13)^2 + (11-13)^2 + (15-13)^2 = 1 + 1 + 0 + 4 + 4 = 10)
    - SSW = 10 + 10 + 10 = 30

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/SSW.png')
new_image = img.resize((250, 100))
display(new_image)
```

- calculate the total sum of squares
    -  total sum of squares = sum of squares between groups + sum of squares within groups
        - SST = SSB + SSW = 153.33
- calculate the degree of freedom
    - degree of freedom between groups
        - dfb = k - 1 = 3 - 1 => 2
    - degree of freedom within groups
        - dfw = N - k = 15 - 3 => 12
    - total degree of freedom
        - dft = dfw + dfb => 14

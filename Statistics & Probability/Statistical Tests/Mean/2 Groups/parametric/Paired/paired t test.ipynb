#### Paired T test:
- used to compare the before and after effect on the continuous variable 
- uses 1 continuous variable, one with before and one with after a condition as a separate column/feature 
- compare the means of two related groups when population standard deviation is unknown and population sample less than <=30

- the dataset is show below
- calculate the difference
- calculate the mean of the difference
    - (5+7+2+5+5+8+3+4+7+5)/10 => 5.1 

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/ttest.png')
new_image = img.resize((450, 400))
display(new_image)
```

- calculate the standard deviation of difference
    - std = 1.97
- calculate the standard error of the mean difference
    - SE = 1.97 / 10 => 0.622
- calculate t statistic
    - t = 5.1 / 0.622  8.20
- calculate Degrees of freedom: 
    - df = 10 - 1 = 9
- calculate p-value
    - using both degree of freedom (9) & t-statistic value
    - p-value() = less than 0.001
    - Since 0.001 < 0.05, the mean of before and after is significantly different. Reject the null hypothesis which denotes there is an effect

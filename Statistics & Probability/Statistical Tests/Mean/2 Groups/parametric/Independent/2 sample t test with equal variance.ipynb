#### 2 Sample T test or Students t test:
- used to compare the means of two groups or variables and find if they are statistically different or not
- use one categorical variable with a continuous variable or use 2 continuous variables

#### Example:
Question:
- Values: 
    - Method A (n1=10): 78, 82, 85, 79, 88, 81, 83, 76, 84, 80
    - Method B (n2=12): 85, 88, 92, 86, 90, 87, 94, 89, 91, 83, 86, 88
- n1 = 10, n2 = 12

Answer:
- A:
    - x` = 81.6
    - std = 3.69

- B:
    - x` = 88.25
    - std = 3.27

Scenario 1 - Assuming Equal Variances:
- calculate the Pooled Variance
    - sₚ² = 12.00834

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/equal_variance.png')
new_image = img.resize((350, 90))
display(new_image)
```

- calculate the standard error:
    - SE = sₚ * √(1/n₁ + 1/n₂) = 1.484
- calculate the t-statistic 
    - t = (x̄₁ - x̄₂) / SE = -6.637

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/2Sample.jpg')
new_image = img.resize((450, 130))
display(new_image)
```

- calculate the degree of freedom:
    - df = n₁ + n₂ - 2 = 10 + 12 - 2 = 20

- calculate the P-Value:
    - using the t-statistic value and degree of freedom 
        - p-value(-6.637) = 0.001
    - p-value(-6.637) is lesser than significance level value (0.05). Hence the mean of Method A and B are statistically different and we reject the null hypothesis

#### Welch T Test:
- the extension of student t test when the variance for both the groups or variables are unequal 

#### Example:
Question:
- Values: 
    - Method A (n1=10): 78, 82, 85, 79, 88, 81, 83, 76, 84, 80
    - Method B (n2=12): 85, 88, 92, 86, 90, 87, 94, 89, 91, 83, 86, 88
- n1 = 10, n2 = 12

Answer:
- A:
    - x` = 81.6
    - std = 3.69

- B:
    - x` = 88.25
    - std = 3.27

Scenario 1 - Assuming UnEqual Variances:
- calculate the t statistic
    - t = (x̄₁ - x̄₂) / √(s₁²/n₁ + s₂²/n₂) => -4.430
- calculate the degrees of freedom:
    - using the formula below, df = 17.8

- calculate the p-value
    - using degree of freedom value and the t-statistic value
    - p-value = <0.001
    - therefore p-value is lesser than significance level value and the mean are of sample A & B are different. Thus, we reject the null hypothesis

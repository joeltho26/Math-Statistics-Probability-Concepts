1. [2 Sample Z test](#2-sample-z-test)

#### 2 Sample Z test:
- used to compare the mean of two groups 
- use with 1 categorical variable consisting of 2 groups with continuous variable or 2 continuous variables from same population
- it is performed when the population standard deviation is known and for both variables sample sizes are greater than >30.

##### Example:
Question:
- Known Population Mean(A) = 500 
- Company A:
    - x` = 1050
    - n = 35
    - σ = 50
- Company B:
    - x` = 1000
    - n = 40
    - σ = 60

Answer:
- Null Hypothesis: μ₁ = μ₂
- Alternate Hypothesis = μ₁ ≠ μ₂
- significance level = 0.05
- Calculate Z-statisitc score
    z-statistic = 3.935
- Determine the P-value
    - two tailed test:
        - 2 * p-value(3.935) is very much lesser than 0.0001
        - Therefore, since p-value is lesser 0.05, its statistically significant i.e. the means of two samples are different. Reject the null hypothesis

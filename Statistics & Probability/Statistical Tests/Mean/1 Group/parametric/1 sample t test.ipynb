#### 1 Sample T test:
- used when the variable is following a distribution
    -  normal, exponential, log-normal, and beta distributions.
- used when we have samples <=30 or when the population standard deviation is unknown
- deals with 1 continuous numerical variable
- used to compare the mean of 1 continuous numerical variable with the hypothesized population mean

##### Example:
Question:
- Hypothesized Mean = 100
- Sample data:
    - 105, 110, 98, 115, 102, 108, 120, 95, 107, 112, 100, 106, 118, 99, 104, 109, 113, 101, 116, 97, 103, 111, 117, 96, 114
- n = 25
- Find the mean, std, t-test value, critical value, significance level, p-value?

Answer:
- Null Hypothesis = 100
- Alternate Hypothesis = >100 (based on the calculated mean)
- mean = 106.5
- std = 8.2
- t-statistic = (106.5 - 100)/1.64 => 3.96
- degree of freedom = n - 1 = 25 - 1 => 24
- significance level = 0.05

Comparing t critical value vs t-statistic:
- to calculate the t critical value take the degree of freedom and significance level value and choose the intercept value from the t distribution table
- two-tailed(when the direction of significance is not known):
    - t critical value = 2.064
- one-tailed (when the direction of significance is known)    
    - t critical value = 1.711 
    - we can see the t-test value greater than one-tailed t critical value i.e. it reject the null hypothesis

Comparing p-value vs significance level:
- p-value is calculated from t distribution table by taking the t-test value and the degree of freedom i.e. the t-test value could be within a range then take the average of it
- one-tailed:
    - p-value(3.96) = 0.0005
    -  since p-value (0.0005) less than significance level value (0.05) then it denotes to reject the null hypothesis and accept the alternate hypothesis since its statistically significant
- two-tailed:
    - p-value(3.96) = 0.001
    - since p-value (0.001) less than significance level value (0.05) then it denotes to reject the null hypothesis and accept the alternate hypothesis since its statistically significant

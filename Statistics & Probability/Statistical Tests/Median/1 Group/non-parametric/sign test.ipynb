#### Sign Test:
- used to compare the median of the sample with the hypothesized or known median
- used with 1 continuous or ordinal variable

##### Example:
Question:
- Values: 155, 160, 148, 152, 158, 165, 145, 151, 156, 162, 153, 159
- n = 12
- Hypothesized median = 150

Answer:
- Null Hypothesis: Median = 150 
- Alternate Hypothesis: Median â‰  150
- calculate the difference between hypothesized median vs actual values
    - the positive difference values are marked with "+" sign
    - the negative difference values marked with "-" sign 
    - the no difference values marked with "=" sign

- count the number of signs "+", "-" & "="
- calculate also the total sign count
    - Number of positive signs (S+) = 10
    - Number of negative signs (S-) = 2
    - Number of ties (equal to 150) = 0
    - Total number of observations (n) = 12
- Calculate the Test statistic:
    - Test Statistic (S) = min(S+, S-) = min(10, 2) = 2
- Calculate P-value:
    - using binomial probability formula: P(k; n, p) = C(n, k) * p^k * (1-p)^(n-k)
    - P(Xâ‰¤2|n=12,p=0.5) = P(X=0) + P(X=1) + P(X=2)
        - P(X=0) = C(12, 0) * (0.5)^0 * (0.5)^12 = 1 * 1 * 0.000244 â‰ˆ 0.0002
        - P(X=1) = C(12, 1) * (0.5)^1 * (0.5)^11 = 12 * 0.5 * 0.000488 â‰ˆ 0.0029
        - P(X=2) = C(12, 2) * (0.5)^2 * (0.5)^10 = 66 * 0.25 * 0.000977 â‰ˆ 0.0161
    - P(X â‰¤ 2) = 0.0002 + 0.0029 + 0.0161 = 0.0192 
    - P-value for two-tailed multiplied by 2: 
        - 2 * 0.0192 = 0.0384
    - if right tailed test, (S+) = 10 will be chosen and P(Xâ‰¥2,|n=12,p=0.5) = P(X=10) + P(X=11) + P(X=12) 
- Since p-value is lesser than significance level value, then the median is statistically significant for the sample i.e. we reject the null hypothesis

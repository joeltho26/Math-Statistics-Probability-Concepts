#### 1 Sample Wilcoxon Signed rank test:
- used to compare the median of the sample distribution with the hypothesized or known median value
- used with continuous or ordinal values 

##### Example:
Question:
- Values: 68, 72, 80, 70, 76, 75, 82, 65, 78, 71
- n = 10
- hypothesized or known median = 75

Answer:
- Null Hypothesis: median = 75
- Alternate Hypothesis: median â‰  75

- Calculate the difference between the observed value and hypothesized or known median and discard the 0 difference value
    - 7, -3, 5, -5, 1, 7, -10, 3, -4 
    - n = 9

- Take absolute values of the difference and rank them from lowest to highest
    - when 2 values have same value, the share the rank by taking the average

    - "+" = 3 + 1 + 4.5 + 2.5 => 11
    - "-" = -4.5 + (-2.5) + (-3) + (-9) + (-6) => -25 => 25 (absolute)
- test statistic value,
    smaller sum value of both "+" & "-" => 11 
- now calculate the critical value (instead of p-value)
    - n = 9 (samples) and significance level value (0.05)
    - critical value => 5
- If test test statistic value (11) > critical value (5), then we fail to reject the null hypothesis

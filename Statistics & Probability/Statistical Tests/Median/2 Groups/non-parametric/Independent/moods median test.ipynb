#### Moods Median Test:
- used to compare the medians of the two groups
- use with 1 categorial variable with two groups vs continuous variable or use 2 samples from same continuous variable from population 

##### Example:
Question:
- Version A:
    - n = 15
- Version B:
    - n = 12

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/mood_data.png')
new_image = img.resize((250, 570))
display(new_image)
```

Answer:
- Null Hypothesis: median of samples same
- Alternate Hypothesis: median of samples not same
- calculate the overall median
    - append the samples one below the other and sort the values
        - 22, 23, 24, 25, 26, 27, 27, 28, 28, 28, 29, 29, 30, 30, 31, 31, 32, 32, 33, 33, 35, 35, 36, 38, 39, 40
    - total sample size combined
        - n = 27
    - median value = (27+1)/2 => 30 seconds
- create a contingency table '

| Website Version | Above Overall Median ( > 30 ) | At or Below Overall Median (  30 ) | Total |
|---|---|---|---|
| A | 4 (31, 33, 35, 32) | 11 (25, 30, 22, 28, 26, 29, 24, 27, 23, 28, 29) | 15 |
| B | 8 (32, 38, 35, 40, 36, 33, 39, 31) | 4 (28, 30, 29, 27) | 12 |
| Total | 12 | 15 | 27 |

- perform chi-squared test of independence
    - calculate expected values
        - Expected (A, Above) = (12 * 15) / 27  6.67
        - Expected (A, Below) = (15 * 12) / 27  8.33
        - Expected (B, Above) = (12 * 12) / 27  5.33
        - Expected (B, Below) = (12 * 15) / 27  6.67
    - calculate chi-squared value
        -  =  [(Observed - Expected) / Expected]
        -  = [(7 - 8.33) / 8.33] + [(8 - 6.67) / 6.67] + [(8 - 6.67) / 6.67] + [(4 - 5.33) / 5.33]
        -   1.06

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/chisqr.jpg')
new_image = img.resize((370, 270))
display(new_image)
```

- calculate the degrees of freedom
    - df = (number of rows - 1) * (number of columns - 1)
    - df = (2 - 1) * (2 - 1) = 1 * 1 = 1
- calculate the p-value
    - using the test statistic value (1.06) and degree of freedom (1)
        - between 0.9 to 0.1 = (0.9 + 0.1)/2 = 0.5
- The p-value (0.5) > significance level (0.05) therefore, we cannot reject the null hypothesis and we can say that there is no statistical difference in the median value between two samples

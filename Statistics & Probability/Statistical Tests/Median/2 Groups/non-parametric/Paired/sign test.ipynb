#### Sign Test:
- used to compare the median of related groups 
- use with 1 continuous variable with before values and after values from a condition

##### Example:
Question:
- sample size (n) = 15

Answer:
- Null Hypothesis: There is no difference in the median before a condition and after a condition
- Alternate Hypothesis: There is difference in the median before a condition and after a condition
- calculate the difference between before and after for the continuous variable
- assign signs "+" for positive values from difference and similarly assign signs "-" for negative values from difference 
    - equal values are omitted

- count the number of signs "+" & "-" and choose the smaller value among both
    - Number of positive signs (+) = 11
    - Number of negative signs (-) = 3 (chosen)
- calculate the p-value using binomial distribution function
    - P(X ≤ 3) = P(X=0) + P(X=1) + P(X=2) + P(X=3) ≈ 0.0001 + 0.0009 + 0.0091 + 0.0349 ≈ 0.045
    - P(X ≥ 11) = P(X=11) + P(X=12) + P(X=13) + P(X=14) ≈ 0.0349 + 0.0091 + 0.0009 + 0.0001 ≈ 0.045
    - The two-tailed p-value = P(X ≤ 3) + P(X ≥ 11) ≈ 0.045 + 0.045 = 0.090
- Hence the p-value (0.090) > significance level (0.05). We fail to reject the null hypothesis

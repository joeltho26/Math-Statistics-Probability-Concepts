#### 2 Sample Wilcoxon Signed Rank Test:
- used to compare median of related group and find the significance after a condition
- uses 1 continuous variable against condition before and after

##### Example:
Question:
- sample size (n) = 10
- significance level = 0.05

Answer:
- Null Hypothesis: There is no effect based on the condition
- Alternate Hypothesis: There is a effect based on the condition

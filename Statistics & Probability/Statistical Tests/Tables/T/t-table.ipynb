#### T-Table:
- provides critical values for performing hypothesis testing and confidence interval calculations
- left scale represents the degree of freedom
- top scale represents the significance level
    - one tail: if we know the distribution well and takes into account of one side of the distribution
        - eg: 0.05 => same
    - two tail: if we dont know the distribution and takes into account of two side of the distribution
        - eg: 0.05/2 => 0.025

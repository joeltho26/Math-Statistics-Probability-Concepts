#### 2 Sample Kolmogorov Smirnov Test:
- used to compare a categorical feature with two groups against 1 continuous variable or compare two samples of continuous values and check if they are of same type of data distribution
- uses ECDF of samples to evalute

##### Examples:
Question:
- Young Adults:
    - Values
        - 210, 225, 195, 230, 215, 200, 220, 235, 190, 240, 205, 218, 228, 202, 212, 232, 198, 222, 217, 208, 227, 238, 193, 219, 223
        - n = 25
- Old Adults:
    - Values: 
        - 250, 265, 235, 270, 255, 240, 260, 275, 230, 280, 245, 258, 268, 242, 252, 272, 238, 262, 257, 248, 267, 278, 233, 259, 263, 247, 269, 277, 237, 253
        - n = 30

Answer:
- Null Hypothesis: same distribution
- Alternate Hypothesis: not same distribution
- calculate the ECDF
    - sort the data in both the samples
    - calculate the frequency of the values for both the samples for each value
    - divide the frequency of the values/total observation for both the samples for each value
- calculate D statisitc
    - maximum difference between the two samples of values
        - D statistic value = 0.440
- calculate the critical value
    - using significance level value (0.05) and number of samples (n=25,n=30)
    - critical value = 0.369

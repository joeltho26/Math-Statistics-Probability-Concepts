#### Wald Wolfowitz Test or 2 Samples Run Test:
- used to check for the distribution of data and check if the both the samples are relevant and from same population with a data distribution or not
- 1 categorical variable with 2 categories and 1 continuous variables or 2 continuous variables from same population data

##### Example:
Question:
- Group A:
    - size (n1) = 6
    - Values = 12,15,18,11,14,16
- Group B:
    - size (n2) = 6
    - Values = 20, 22, 25, 19, 23, 21

Answer:
- Null Hypothesis: the dsitribution is same for Group A & B
- Alternate Hypothesis: the dsitribution is not same for Group A & B

| Group A | Group B |
| --- | --- |
| 12 | 20 |
| 15 | 22 |
| 18 | 25 |
| 11 | 19 |
| 14 | 23 |
| 16 | 21 |

- calculate the p-value 
    - using the z-statistic value
        - P-value(3.028) = (0.0012 + 0.0013)/2 = 0.00125
        - Two tailed test:
            - P-value = 2 * 0.00125 => 0.0025
- Hence p-value (0.0025) < significance level (0.05). Reject the null hypothesis and say they both have different distributions

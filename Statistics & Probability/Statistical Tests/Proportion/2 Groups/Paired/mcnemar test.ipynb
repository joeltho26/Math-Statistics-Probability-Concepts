#### Mcnemar Test:
- used to compare the median of the sample before and after a condition and find it significantly different
- use 1 binary variable with before and after a condition

##### Example:
Question:
sample size = 100 consumers

Answer:
- Null Hypothesis: proportion of binary variable before and after the condition is the same
- Alternate Hypothesis: proportion of binary variable before and after the condition is not the same

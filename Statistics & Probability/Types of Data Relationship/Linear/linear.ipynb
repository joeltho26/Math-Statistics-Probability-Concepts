#### Linear Relationship:
- assocated with constant rate of change and moves either forward or backward direction
- distinguished based of strength and direction
- Based on direction, there are two types:
    - positive linear relationship
    - negative linear relationship
- Based on strength, there are two types:
    - strong linear relationship
    - weak linear relationship

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/linear.png')
new_image = img.resize((750, 550))
display(new_image)
```

Positive Linear Relationship:
- when one variable increase, there is increase in another variable with constant rate of change 

Negative Linear Relationship:
- when one variable increase, there is decrease in another variable with constant rate of change

Strong Linear Relationship:
- when the datapoints for change in variable 1 to change in variable 2 forms almost a straight line

Weak Linear Relationship:
- when the datapoints for change in variable 1 to change in variable 2 are scattered around

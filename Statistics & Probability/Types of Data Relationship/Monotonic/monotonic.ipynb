#### Monotonic Relationship:
- ```not associated``` with ```constant rate of change``` 
- relationship move only in ```1 direction``` wither forward or backward
- linear relationship and some non linear relationship is a type of monotonic relationship
- it is associated with ```direction``` but ```not magnitude``` 
    - +ve or -ve monotonic relationship

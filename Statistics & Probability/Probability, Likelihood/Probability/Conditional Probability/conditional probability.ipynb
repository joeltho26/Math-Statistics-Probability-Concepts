##### Eg 2: Candy and Soda:
- Loves Candy only = 6; Loves Soda only = 7; Loves Soda and Candy = 2; Doesnt love Candy and Soda = 3

Soln: 
- Creating contingency table

|  | Loves Candy | Doesnt Love Candy |
| --- | --- | --- |
| Loves Soda | 2 | 5 |
| Doesnt Love Soda | 4 | 3 |


- Now calculating the probabilities
    - P (Loves Soda & Candy) = 2/14
    - P (Loves Soda & Doesnt Love Candy) = 5/14
    - P (Loves Candy & Doesnt Love Soda) = 4/14
    - P (Doesnt Love Candy & Doesnt Love Soda) = 3/14

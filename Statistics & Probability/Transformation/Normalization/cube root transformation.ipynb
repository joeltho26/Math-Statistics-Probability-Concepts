#### Cube Root Transformation - Gaussian:
- used to achieve normality of data
- used for non gaussian distributed data
- applies cube root to the values of a variable
- used to reduce the impact of outliers
- used to reduce skewness and kurtosis
- used to reduce right skewed data
- used for both positive and negative values

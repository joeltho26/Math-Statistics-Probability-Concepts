#### Box-Cox Transformation:
- used to convert non gaussian data into gaussian distribution
- most commonly used to reduce right skew
- reduces skewness and kurtosis of data
- reduces the impact of outliers
- λ is the transformation factor and it ranges between +5 to -5
- it works only with positive values and if zero or negative values is present it results in error
- Based on λ value:
    - λ = 1 => no transformation
    - λ = 0 => log base e transformation
    - λ = 0.5 => square root transformation
    - λ = -0.5 => square root & reciprocal transformation
    - λ = -1 => reciprocal transformation

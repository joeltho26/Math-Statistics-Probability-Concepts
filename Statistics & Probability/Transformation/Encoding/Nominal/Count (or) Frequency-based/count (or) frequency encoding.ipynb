#### Count (or) Frequency-based Encoding:
- calculate the ```count``` of each label
- calculate the ```probability``` of each label
    - ```divide``` the ```count``` of each label with the ```overall count``` (rows) of dataset
- if ```2``` labels having ```same frequency```, 
    
    - ```lexicographical order```
    
    - ```ordered mean encoding``` (alternate approach)
    
    - ```introduce random noise```

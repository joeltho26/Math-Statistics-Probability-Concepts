#### One Hot Encoding:
- used to encode the ```categorical``` values 
- encoding creates ```new columns``` of data
- used when ```small``` number of categories present in the feature/variable
- if used for encoding variables with ```large``` number of ```categories``` result in ```curse of dimensionality```
- the ```length``` of the encoded value is ```N```
    - ```N``` = number of distinct labels in the categorical variable
    - ```each label``` of the categorical variable ```creates``` a ```new column```
- experiences ```dummy variable trap```, thereby ```remove one label``` as ```reference``` to encode

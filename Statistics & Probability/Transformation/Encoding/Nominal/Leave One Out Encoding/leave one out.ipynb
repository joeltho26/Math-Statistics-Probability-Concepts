#### Leave One Out Encoding:
- encoding is performed based on ```each row```
- it helps in overcoming ```data leakage```
    - ```leaking information``` about the ```output class``` to the ```encoded values``` during transformation of categorical data into numerical values
- ```Train``` set:
    - while ```encoding``` for ```category label``` of the ```current row```, the current row ```values``` are ```excluded``` both in numerator and denominator
- ```Test``` set:
    - the encoding of labels is based on the ```probability``` of ```each category label``` w.r.t ```output class labels``` based on ```training dataset```

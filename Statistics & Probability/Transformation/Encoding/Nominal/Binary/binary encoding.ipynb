#### Binary Encoding:
- used to encode categorical value in ```binary format```
    - eg: "High" => "010"
- the encoding is represented in ```1``` and ```0```
- assigns a ```distinct integer``` to each category in the variable and converted into ```binary representation```
    - each ```bits create``` a ```new column```

#### Max Absolute Scaler:
- it is used the scale the data between the range -1 to + 1
- the ratio of value by maximum absolute value
- if distribution is unknown or not gaussian distributed
- sensitive to outliers
- useful for sparse data
- handles both positive and negative values
- used for algorithms that support,
    - recommendation systems
    - TF-IDF (NLP)
    - Neural Network
    - Image Processing

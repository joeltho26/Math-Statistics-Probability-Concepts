#### Standard Scaler:
- it is used to rescale the values between -3 to + 3
- it is called as standardization
- standardization is evaluated the same way as z-score
- the ratio of difference between the value & the mean value by standard deviation of the data
- used when data is gaussian distributed
- sensitive to outliers

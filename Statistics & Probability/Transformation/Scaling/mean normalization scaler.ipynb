#### Mean Normalization:
- it is used to rescale the values between - 1 to 1
- the ratio of difference between the value and mean value by difference between maximum value and minimum value
- works for both gaussian or non gaussian distribution
- sensitive to outliers
- used when we want the values to center around the mean
- works with both positive and negative values
- used with algorithms that work with calculating distances
    - ANN
    - image processing
    - Linear models

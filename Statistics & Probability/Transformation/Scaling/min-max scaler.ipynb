#### Min-Max Scaler:
- used to scale data between 0 - 1 (only positive values) or -1 to 1 (both positive and negative values)
- the ratio of difference between the value & the minimum value by difference between maximum value & minimum value
- works for both gaussian or non gaussian distribution
- sensitive to outliers
- works with both positive and negative values
- it is commonly used in algorithms that are based on distance calculations,
    - Neural Networks
    - Support Vector Machine
    - KNN
    - Image processing

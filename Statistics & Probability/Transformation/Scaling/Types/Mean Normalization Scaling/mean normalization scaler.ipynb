#### Mean Normalization:
- ```rescale``` the values between ```-1``` to ```1```
- preferred for ```non gaussian``` or ```unknown``` distribution of data
- ```sensitive``` to outliers
- used when we want the values to ```center around mean```
- used with algorithms that work with calculating distances
    - ANN
    - image processing
    - Linear models

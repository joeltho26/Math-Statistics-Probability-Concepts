#### Standard Scaler:
- ```scales``` the values between ```-3``` to ```+3```
- it is called as ```standardization```
    - transforms the data to exhibit distribution properties ```μ = 0``` and ```σ = 1``` 
- standardization same as ```z-score```
- preferred for ```gaussian``` distribution of data 
- ```sensitive``` to outliers

Note:
- if a data distribution has ```μ = 0``` and ```σ = 1``` doesn't mean it follows normality

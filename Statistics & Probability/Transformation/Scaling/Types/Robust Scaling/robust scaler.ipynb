#### Robust Scaler:
- they are ```robust``` to ```outliers```
- rescales the data
- transforms the data ```without changing``` the ```shape``` of distribution and ```relationship``` between variables
- preferred on ```non gaussian``` distributed data or distribution is ```unknown```
- ```Median``` & ```IQR``` values used to perform re-scaling 
    - IQR = ```Q3``` (75th percentile) - ```Q1``` (25th percentile)

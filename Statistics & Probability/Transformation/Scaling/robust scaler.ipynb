#### Robust Scaler:
- they are robust to outliers
- mostly used for non gaussian distributed data or if the distribution is unknown
    - can be also used for gaussian distributed data but not recommended
- robust scaler range between 25th percentile - 75th percentile
- the ratio of difference between the value & median value by IQR 
    - IQR = Q3 (75th percentile) - Q1 (25th percentile)
- works with positive and negative values

#### Gaussian Transformation Methods:
- transforms the data by ```changing``` the ```shape``` of distribution and ```relationship``` between variables
- transforms data to be ```normally distributed```
- useful when ml algorithms that require the data to be normally distributed
- reduces the impact of ```outliers```
- Types of Gaussian Transformation:
    
    - ```Log Transformation```

    - ```Reciprocal Transformation```

    - ```Square root Transformation```

    - ```Cube root Transformation```

    - ```Exponential Transformation```

    - ```Box-Cox Transformation```

    - ```Yeo-Johnson Transformation```

    - ```Quantile Transformation```

    - ```Power Transformation```

Note:
- normality of data achieved

    - ```skew = 0```

    - ```kurtosis = 3```

    - ```mean``` need not be 0

    - ```std``` need not be 1

- standard normality:

    - ```skew = 0```

    - ```kurtosis = 3```

    - ```mean = 0```

    - ```std = 1```

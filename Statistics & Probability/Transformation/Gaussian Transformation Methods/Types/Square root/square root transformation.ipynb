#### Square Root Tramsformation - Gaussian:
- achieve ```normality``` of data i.e. ```gaussian``` distribution
- reduces ```skew``` and ```kurtosis``` of data
- reduces impact of ```outliers```
- works best for ```right/+ve``` skewed data
- works with ```+ve``` and ```0``` values
- ```-ve``` values in square root transformation results in ```imaginary``` numbers
    - to overcome, add a ```small constant```

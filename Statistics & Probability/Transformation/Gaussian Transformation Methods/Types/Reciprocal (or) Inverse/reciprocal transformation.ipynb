#### Reciprocal or Inverse Transformation:
- reduce the ```skew``` and ```kurtosis```
- transforms the data into ```gaussian``` distribution
- reduce the impact of ```outliers```
- used for ```right/+ve``` skewed data
- works only for ```+ve``` & ```-ve``` values, 
- ```0``` values are handled by adding ```small constant```

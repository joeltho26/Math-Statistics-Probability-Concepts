#### Cube Root Transformation:
- achieves ```normality``` of data
- transforms data into ```gaussian``` distributed data
- used to reduce the impact of ```outliers```
- reduce ```skew``` and ```kurtosis```
- reduce ```right/+ve``` skewed data
- works with ```+ve```, ```-ve``` and ```0``` values

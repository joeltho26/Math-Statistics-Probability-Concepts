#### Log Transformation:
- transforms data into ```gaussian``` distribution
- used for the ```right/+ve``` skewed data
- reduces ```skew``` and ```kurtosis``` of data
- ```log base e``` is applied to the data for transformation
- only works with ```+ve``` values
- ```-ve``` and ```0``` values are handles by,
    - we add a ```small constant``` value
    - taking the ```absolute``` of value
        - eg: -3.5 => 3.5
- reduces the impact of ```outliers```

#### Linear Transformation:
- have ```fixed origin``` at ```0```
- ```parallel and evenly spaced straight grid lines```
- the ```diagonal``` should be also a ```line``` and not curvy
- transformations

    - ```rotation``` (around origin)

    - ```reflection```

    - ```scaling``` (dilations)

    - ```shears```

Note: 
- based on ```transformed i^``` and ```j ^ basis vectors``` we can calculate ```transformed vector```
    - the ```scalar values``` for the ```basis vector``` remains the ```same before``` and ```after``` the transformation

example 1: ```2x2 transformation matrix``` shown below holds the ```transformed i^``` and ```j^``` position which tells us about the ```transformed vector```

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/transformation-matrix.png')
new_image = img.resize((150, 120))
display(new_image)

img = Image.open('images/transformation-matrix1.png')
new_image = img.resize((350, 210))
display(new_image)
````

Note:
- if the ```transformed i^``` & ```j^ basis vector``` becomes ```linearly dependent```
    - 2D: the transformed i^ & j^ basis vector will move only across the line
    - 3D: the transformed i^ & j^ basis vector will move only across the plane of x & y axes

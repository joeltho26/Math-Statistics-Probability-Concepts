#### Scalar Multiplication:
- multiplying a scalar value with a vector result in same vector ```expand``` or ```shrink```
    - when multiplying with ```-ve``` scalar value, the vectors get flipped onto ```opposite``` direction
- also called as scaling

````python

from PIL import Image 
from IPython.display import display

img = Image.open('images/scalar1.png')
new_image = img.resize((65, 40))
display(new_image)

img = Image.open('images/scalar.png')
new_image = img.resize((220, 70))
display(new_image)

img = Image.open('images/scalar2.png')
new_image = img.resize((230, 150))
display(new_image)

img = Image.open('images/scalar3.png')
new_image = img.resize((380, 160))
display(new_image)
````

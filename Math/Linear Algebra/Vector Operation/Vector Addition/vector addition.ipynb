#### Vector Addition:
- adding 2 vectors results in a new vector with `same dimension`
- Visually adding 2 vectors,
    - take one vector `w→` and place the vector's `tail` onto the `tip` vector `v→` and then draw a `straight line` joining the `tip` of vector `v→` from `origin`
    - `direction of movement` is considered i.e. `w→` vector is moved onto `v→`

#### Cross Product:
- conceptualized with 3 dimensions
- ```cross product``` of ```2 2D vectors``` in 3D space result in a ```new vector perpendicular``` to ```x & y axis i.e. z-axis``` and also ```scalar``` quantity i.e. ```length of the vector``` (area of the parallelogram)
- ```cross product``` of ```2 3D vectors``` result in a ```new 3D vector orthogonal``` to the ```plane``` containing 2 vectors
    - to identify the ```direction``` of ```new vector``` from plane containing 2 vectors using flemming's right hand rule
- we use ```determinants``` to find the ```cross product of 2 2D/3D vectors```
- the ```order of vectors``` in cross product is ```important```
    - i.e. ```orientation``` is represented by ```-ve sign```

- taking the cross product of 3 3D vectors will result in calculating the volume of the parallelogram
- in 3D cross product, we have 2 fixed vectors and 1 varying vector
- the volume is given by the scalar triple product formula
    - resulting in a new vector which constitutes the volume by combining the values of three vectors

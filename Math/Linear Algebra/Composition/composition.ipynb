#### Composition:
- a single transformation matrix holds the multiplying 2 transformation matrix
    - a single ```matrix A``` = ```rotation``` (transformation 1) + ```shear``` (transformation 2)
    - A = M1 x M2

Note:
- ```order of matrix multiplication``` is ```followed```
    - ```M1 x M2 â‰  M2 x M1```
- ```Associative```:
    - ```(M1 x M2) x M3 = M1 x (M2 x M3)```
- transformation matrix 2x2 and a 2D vector, results in transformed 2D vector
- transformation matrix 3x3 and a 3D vector, results in transformed 3D vector

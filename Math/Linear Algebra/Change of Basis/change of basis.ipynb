#### Change of Basis:
- changing the basis vectors of a vector space to a different set of basis vectors
- Process:
    - take vector in Jen's coordinate system & multiply with change of basis matrix (basis vector of Mark's w.r.t Jen's) to find the corresponding vector in Marks's coordinate system 

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/basis.png')
new_image = img.resize((880, 485))
display(new_image)
````

- we can perform transformation in one coordinates and then convert it back to any coordinate system
    - Process:
        - we first changes the basis from Jen's to Mark's coordinate system 
            - using change of basis matrix
        - we then performed transformation 
            - using transformation matrix
        - we then revert the basis from Marks's to Jen's coordinate system as shown in the image below
            - using the inverse of change of basis vector
    - eg: ```AMAv = w```
        - A = inverse of change of basis matrix
        - A = change of basis matrix
        - M = transformation matrix
        - v = vector in Jen's coordinate system
        - w = transformed vector in Jen's coordinate system

#### Inverse Matrix:
- reverse the transformation 
- represented by A⁻¹
- ```transformation``` + ```inverse transformation``` = No transformation
    - ```A A⁻¹ = I``` (Identity matrix)
- in linear system equations, we apply inverse transformation to find the actual vector using transformed vector
    -  ```A⁻¹V→ = X→```
- Inverse matrix ```A⁻¹``` exists when there is det ≠ 0
    - i.e. when transformation result in ```dimension reduction``` eg: 3D space squishing into 2D plane, line or a point, inverse matrix not available

Calculating the Inverse Matrix:
- calculate the determinant of matrix
- calculate the adjoint of matrix:
    - adjoint of a matrix, also known as the adjugate
    - is the transpose of its cofactor matrix
    - finds the cofactor of each element in the original matrix
    - arranging these cofactors into a new matrix (the cofactor matrix)
    - then swapping the rows and columns of this new matrix

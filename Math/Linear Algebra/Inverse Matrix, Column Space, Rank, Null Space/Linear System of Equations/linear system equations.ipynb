#### Linear System of Equations:
- by re-arranging the equation, tells us about ```X→``` after transformation (```matrix A```) will result in vector ```V→``` (transformed vector)
    - ```A X→ = V→```
- the ```linear equation``` can be solved by inverse matrix ```A⁻¹``` and transformed vector ```V→``` to identify the original vector ```X→```
    - ```A⁻¹ V→ = X→```

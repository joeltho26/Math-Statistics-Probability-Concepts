#### Rank:
- the values of rank tells us about the ```output dimensions/column space``` from transformation
    - ```Rank = 1```: transformation in 3D space results in a ```line```
    - ```Rank = 2```: transformation in 3D space results in a ```plane```
    - ```Rank = 3```: transformation in 3D space results in a ```full volume```
        - when we see the ```rank value after transformation equal``` to the number of ```original dimension before transformation``` is called ```full rank```

#### Cramer's Rule:
- used to ```calculate``` the ```linear system equations``` 
    - finding values for variables: X₁, X₂
- it uses ```determinants``` to perform the calculations
- it is considered ```slower``` than ```gaussian elimination``` and ```row echelon```

Note:
- After transformations,
    - dot product of 2 vectors will change

        ```(X₁→) . (X₂→) ≠ (V₁→) . (V₂→)```

        - X₁, X₂ => original vector
        - V₁, V₂ => transformed vector
- Orthonormal transformations are ones that preserve the dot product even after transformations
    - basis vectors are orthonormal

    ```(X₁→) . (X₂→) = (V₁→) . (V₂→)```

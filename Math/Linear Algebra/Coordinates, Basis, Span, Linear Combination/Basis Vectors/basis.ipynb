#### Basis Vectors:
- in coordinate system x & y, there are 2 basis vectors ```i^``` & ```j^```
    - ```i^``` & ```j^``` are of unit length

- example of an vector with i and j basis vector 
    - 3i + (-2)j

Note:
- the ```basis``` of ```vector space``` are a set of ```vectors``` that are ```linearly independent``` that ```span``` the ```whole space```

#### Span:
- ```linear combination``` of ```'v'``` and ```'w'``` vector with ```all possible``` values, tells about the ```span``` of ```'v'``` & ```'w'``` vectors
- span talks about the ```accessible range``` in the ```N dimensional space```

```example 1```: 2 vectors in 2D space

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/linear-combination2.png')
new_image = img.resize((520, 265))
display(new_image)
````

```example 2```: 2 vectors in 3D space:
- the span of the vectors would be a plane in x & y axes

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/linear-combination3.png')
new_image = img.resize((400, 365))
display(new_image)
````

```example 3```: 3 vectors in 3D space ensuring all the ```vectors 'v', 'w', 'c'``` are ```orthogonal``` (90 degree apart)
- span is the whole cuboid in x,y & z
- vectors separated by 90 degree are called ```linearly independent```

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/span3.png')
new_image = img.resize((300, 205))
display(new_image)
````

```example 4```: 2 vectors ```'v'``` & ```'w'``` in ```3D space``` are ```orthogonal``` but 3rd vector ```'c'``` is ```parallel``` to ```one``` of the ```vectors``` ```'v'``` or ```'w'```
- vectors are ```linearly dependent```, then we ```omit one``` of the ```vector``` that ```overlaps``` in the same plane axis.

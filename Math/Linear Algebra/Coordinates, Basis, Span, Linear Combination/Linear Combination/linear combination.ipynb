#### Linear Combination:
- process of ```scaling one``` or ```more vectors``` and ```adding them together``` to form a ```new vector```
- example: ```(a,b) vector = [1.49 0.84]``` i.e. ```1.49v + 0.84w```  is a linear combination of basis vectors ```v→``` and ```w→```

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/linear-combination.png')
new_image = img.resize((150, 50))
display(new_image)

img = Image.open('images/linear-combination1.png')
new_image = img.resize((220, 160))
display(new_image)
````

#### Co-ordinate System:
- the co-ordinate system for,
    - 2 dimensions: x and y axis
    - 3 dimensions: x,y & z axis 
- the coordinates axes ```x``` & ```y``` are ```orthogonal``` and ```intersect``` at a point called ```origin```

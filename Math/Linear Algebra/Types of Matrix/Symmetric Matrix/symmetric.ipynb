#### Symmetric Matrix:
- a `matrix A` when `transposed` produces the `same matrix A`
    - elements moved from rows to columns and columns to rows produces the same matrix
    - `A = Aáµ€`
- elements are `mirrored` across the `main diagonal`
- `square matrix`

Properties:
- `eigen` values are `real numbers`
- `sum` and `product` of symmetric matrices are `symmetric`
- a symmetric matrix is `diagonalizable` i.e. `diagonal matrix` with `eigen` values on its `diagonals` 
- symmetric matrix can be `diagonalized` by an `orthogonal matrix`

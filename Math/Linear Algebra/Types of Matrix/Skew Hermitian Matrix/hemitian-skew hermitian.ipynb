Skew Hermitian Matrix:
- `square matrix`
- the `diagonal elements` are `0` 
- `conjugate` and `transpose` the matrix to get the `-ve` of the matrix`
    - `A = -Aᴴ`
- `non diagonal elements` are `complex numbers`
    - complex number:
        - `3-2i`

`Properties`:
- the `eigenvalues` are `0` or `imaginary numbers`
- `eigenvectors` corresponding to distinct `eigenvalues` are `orthogonal`
- the `sum` of `2 Skew Hermitian matrices` is `Skew Hermitian`
- `product` of `2 Skew Hermitian matrices`, 
    - results in a `Hermitian matrix` only if they are commute `AB=BA`. 
    - results in `not Hermitian matrix` if they are not commute `AB≠BA`.
- `trace` of Skew Hermitian matrix is always `0` or `purely imaginary number`
- `determinant` of Skew Hermitian matrix is always `0` or `purely imaginary number`
- `inverse` of Skew Hermitian Matrix = `Skew Hermitian matrix`
- `conjugate matrix` of a Hermitian matrix is `skew symmetric`
- if A is a `Skew Hermitian matrix` of `any order` and `k` is a real scalar, then `kA` is also a Hermitian matrix as `(kA)ᴴ = kAᴴ = -kA`
- if A is a `skew-Hermitian matrix` of any order, then An is: `Skew-Hermitian` for `odd` n and `Hermitian` for `even` n

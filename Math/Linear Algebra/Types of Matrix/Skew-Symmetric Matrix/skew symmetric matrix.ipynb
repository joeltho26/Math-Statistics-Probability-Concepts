#### Skew Symmetric Matrix:
- a `matrix A` when `transposed` produces the `-ve` of same matrix `A`
    - elements moved from rows to columns and columns to rows produces the -ve of the same matrix 
    - `A = Aáµ€`
- elements are `mirrored` across the main diagonal
- the `prinicpal diagonal elements` contain `0`
- `square matrix`

`Properties`:
- `eigenvalues` of a skew-symmetric matrix are either `0` or purely `imaginary`
- `determinant` of a skew-symmetric matrix
    - if `size` of the matrix is `odd` = `0`
    - if `size` of the matrix is `even`, `product` of the `eigenvalues`
- `eigenvectors` corresponding to distinct `non-zero eigenvalues` of a skew-symmetric matrix are `orthogonal`.

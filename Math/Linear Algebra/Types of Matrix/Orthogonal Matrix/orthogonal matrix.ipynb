#### Orthogonal Matrix:
- An orthogonal matrix when multiplied by its transpose, results in the identity matrix
    - ```AAᵀ = AᵀA = I```
    - ```Aᵀ = A⁻¹```
- is a ```square``` matrix
- consist of ```orthonormal rows and columns```,
    - vectors in a matrix that are both orthonormal (perpendicular) to each other and have a unit length (magnitude of 1)
    - ```dot product``` of any two different ```row or column vectors``` is ```0```

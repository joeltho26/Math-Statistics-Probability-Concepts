#### Hermitian Matrix:
- `square matrix`
- `conjugate` and `transpose` of matrix A will result in the `same original matrix A`
    - `A = Aᴴ`
- the `principal diagonal elements` are `real numbers`
- `non diagonal elements` are `complex numbers`

Properties:

- the `eigenvalues` are `real numbers`
- `eigenvectors` corresponding to distinct `eigenvalues` are `orthogonal`
- the `sum` of `2 Hermitian matrices` is `Hermitian`
- `product` of `2 Hermitian matrices`, 
    - results in a `Hermitian matrix` only if they are commute `AB=BA`. 
    - results in `not Hermitian matrix` if they are not commute `AB≠BA`.
- `trace` of Hermitian matrix is always real number
- `determinant` of Hermitian matrix is always real number
- `inverse` of Hermitian Matrix = Hermitian matrix
- `conjugate matrix` of a Hermitian matrix is `Hermitian`
- if A is a `Hermitian matrix` of `any order` and `k` is a `real scalar`, then kA is also a Hermitian matrix as `(kA)ᴴ = kAᴴ = kA`

Note:
- `commute` means the `any order of product operation` results in `same output`

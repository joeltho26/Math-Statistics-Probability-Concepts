- used in eigenvalue and eigenvector analysis
- mathematical operation:
    - diagonal matrix multiplies a vector:

    - matrix multiplication:
        - multiply the corresponding diagonal elements
    - Matrix inversion:
        - reciprocal of each diagonal element

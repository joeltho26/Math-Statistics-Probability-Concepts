#### Upper Triangular Matrix:
- a `square` matrix
- elements `below` the main diagonal elements are `0`

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/upper.png')
new_image = img.resize((235, 210))
display(new_image)
```

`Strict Upper Triangular Matrix`:
- the lower element from the main diagonal and also including the main diagonal are `0`

`Unit Upper Triangular Matrix`:
- the lower element from the main diagonal are 0 and the main diagonal values are `1`

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/strict_upper.png')
new_image = img.resize((205, 110))
display(new_image)

img = Image.open('images/unit_upper.png')
new_image = img.resize((205, 110))
display(new_image)
```

`Properties`:
- `transpose` of `upper` triangular matrix = `lower` triangular matrix
- `determinant` of triangular matrix = `product` of `diagonal elements`
- `inverse` of upper triangular matrix = upper triangular matrix
- triangular matrix is `invertible` only if all elements of the `principal` `diagonal` are not `0`
- `product` & `sum` of upper triangular matric = upper triangular matrix

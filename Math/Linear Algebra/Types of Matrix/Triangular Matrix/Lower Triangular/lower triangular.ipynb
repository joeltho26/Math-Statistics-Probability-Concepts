#### Lower Triangular Matrix:
- a `square` matrix
- elements `below` the `main diagonal elements` are `0`

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/lower.png')
new_image = img.resize((235, 190))
display(new_image)
```

`Strict Lower Triangular Matrix`:
- the upper element from the main diagonal and also including the main diagonal are `0`

`Unit Lower Triangular Matrix`:
- the upper element from the main diagonal are 0 and the main diagonal values are `1`

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/strict_lower.png')
new_image = img.resize((205, 120))
display(new_image)

img = Image.open('images/unit_lower.png')
new_image = img.resize((205, 110))
display(new_image)

```

`Properties`:
- `transpose` of lower triangular matrix = `upper` triangular matrix
- `determinant` of triangular matrix = `product` of `diagonal elements`
- `inverse` of lower triangular matrix = `lower triangular matrix`
- triangular matrix is `invertible` only if all elements of the `principal diagonal` are `not 0`
- `product` & `sum` of lower triangular matric = `lower triangular matrix`

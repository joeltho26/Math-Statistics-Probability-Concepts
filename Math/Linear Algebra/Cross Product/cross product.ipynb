#### Cross Product:
- ```cross product``` of ```2 2D vectors``` result in a ```scalar``` quantity i.e. ```area of the parallelogram``` of 2 vectors (magnitude)
    - we use ```determinants``` to find the ```cross product of 2 2D vectors```
- the ```order of vectors``` in cross product is ```important```
    - i.e. ```orientation``` is represented by ```-ve sign```
- ```cross product``` of ```2 3D vectors``` result in a ```new 3D vector perpendicular``` to the ```area of parallelogram``` of 2 vectors
    - to identify the ```direction``` of ```new vector``` from parallelogram using flemming's right hand rule
    - we use ```determinants``` to find the ```cross product of 2 3D vectors```

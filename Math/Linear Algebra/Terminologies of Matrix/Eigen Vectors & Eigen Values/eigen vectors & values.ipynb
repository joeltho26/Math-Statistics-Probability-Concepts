#### Eigen Vector and Eigen Values:
- `vectors after transformation` still lie on the `same span` as their initial ones but can `stretch` or `squish` based on the `eigen value`
- can have `-ve` eigen value i.e. `orientation` eg: -(1/2), i.e. it can be squished and flipped but the span remains the same
- always defined for `square matrices`
- `real number` and `not imaginary numbers`
    - if eigen value is `imaginary numbers` means there are `no eigen vectors`

Note:
- In `3D rotation`:
    - eigen vector for 3D rotation i.e. span remains same after rotation, called ```axis of rotation```

To calculate the Eigen Value and Eigen Vector:
- `Av = v` => `(A-I)v=0`
    - we add the identity matrix because we cannot subtract a matrix with a scalar

Note:
- since `(A-I)` is not `invertible`, then determinant `|A-I| = 0` used to calculate the eigen value
- to calculate the eigen vector, substitute eigen value `` into `(A-I)v=0`
    - after identifying the relationship between variables, we assume a sample value for one of the variable and evaulate the eigen vectors
    - verify the equation by substituting eigen value and vector `(A-I)v=0`

#### Rank:
- tells us about the ```dimensions of column space``` from transformation
    - ```Rank = 1```: transformation in 3D space results in a ```line```
    - ```Rank = 2```: transformation in 3D space results in a ```plane```
    - ```Rank = 3```: transformation in 3D space results in a ```full volume```
        
Note:
- `Full Rank`: rank value remains the `same before` & `after` transformation 

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rank1.png')
new_image = img.resize((130, 120))
display(new_image)

img = Image.open('images/rank2.png')
new_image = img.resize((230, 215))
display(new_image)

img = Image.open('images/rank3.png')
new_image = img.resize((230, 215))
display(new_image)
````

`Calculate the Rank of a Matrix`:

Property:
- the value of `rank` will `<=` `smallest` value of either `row` (m) or `columns` (n)
    - i.e. `rank(A) <= min(m, n)`
- The `rank` of a matrix `=` to the number of `leading 1's` (pivot) in its `row echelon form` (REF) or `reduced row echelon form` (RREF)

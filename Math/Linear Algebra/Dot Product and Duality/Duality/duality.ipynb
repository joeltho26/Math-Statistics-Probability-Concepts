#### Duality:
- when two different operations have correspondence such as below
    - ```vector matrix (1x2) product = dot product of 2 vectors```

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/duality.png')
new_image = img.resize((600, 240))
display(new_image)

img = Image.open('images/duality1.png')
new_image = img.resize((350, 200))
display(new_image)
````

Note:
- when applying linear transformation, 
    - scenarios like, 
        - ```1x2 matrix with 2D vector 2x1 => scalar```
            - the matrix represents the transformed ```i^``` and ```j^```
                - ```row x cols``` as shown in image above
            - dot product of 2 vectors is also the same
            - it means that, when transformation of ```2D space``` results into a ```line```
                - there is an ```associated vector``` i.e. ```transpose``` of the ```transformation matrix```
        - ```1x3 matrix with 3D vector 3x1 => scalar```, same as 2D vectors explanation

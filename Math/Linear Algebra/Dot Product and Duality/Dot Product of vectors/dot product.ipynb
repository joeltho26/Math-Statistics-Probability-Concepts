#### Dot Product of vectors:
- ```dot product``` of 2 or 3 vectors result in ```scalar``` value
- ```dot product of 2 vectors``` is generally considered as ```transformation of a vector```
- when performing a dot product, vectors must have the ```same number of dimensions``` 
- Visually,
    - while performing dot product, one of the ```vector w→``` is ```projected``` onto the ```other vector v→``` and then scaled by ```w→``` value
    - here the order of projecting 1 vector onto the other doesnt matter

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/dot.png')
new_image = img.resize((320, 180))
display(new_image)

img = Image.open('images/dot2.png')
new_image = img.resize((690, 280))
display(new_image)

img = Image.open('images/dot3.png')
new_image = img.resize((610, 280))
display(new_image)
````

Secnarios of dot products:
- if both vectors are in ```same``` direction, ```dot product > 0```
- if both vectors are in ```opposite``` direction, ```dot product < 0```
- if both vectors are ```orthogonal``` to each other, ```dot product = 0```

#### Determinant:
- after ```transformation of vector```, the area can ```increase```, ```decrease```, or still be the ```same```
- tellls us about the ```factor of change``` in ```area of basis vector i^``` and ```j^``` after transformation in terms of unit vector scale is represented in ```magnitude``` 
    - ```det(A) = 3``` (factor of change in ```magnitude```)
- ```-ve sign``` in ```determinant``` means ```orientation```
    - consider flemming left hand rule and after orientation consider the flemming right hand rule
- in 3D dimension, it explains the ```volume of change``` after transformation (shape called ```parallelepiped```)


Note:
- ```factor of change of area``` in ```1 grid square``` will be the ```same``` for ```other grid``` as per the ```property``` of ```linear transformation```
- if ```transformation``` squishes the basis vectors onto ```overlapping``` line, then ```determinant = 0```

- to calculate the determinants
    - last image shows the calculations for determinants in 3D
- Scenarios for various det values: (initial A = 1)
    - ```det(A) > 1```: expand
    - ```det(A) < 1```: shrink
    - ```det(A) = 1```: change shape not area
    - ```det(A) = 0```: collapsing into a line or a single point
    - ```det(A) < 0```: orientation
- ```det(M1M2) = det(M1) det(M2)```

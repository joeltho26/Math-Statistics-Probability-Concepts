#### Logarithm:
- it is used to reduce the wide ranging values and make it substantially closer to each other
- 3 types of logarithms
    - Log base 10:
        - known as common logarithm
        - represented as Log base 10 or Log
        - used when dealing with number system
    - Log base e:
        - value of e = 2.718
        - known as natural logarithm
        - represented as Log base e or ln
        - used when we find exponential growth or decay
    - Log base 2
        - known as binary algorithm
        - used when representing binary information

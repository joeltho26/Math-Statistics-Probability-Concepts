```python
class Hash:
    def __init__(self,size):
        self.size = size
        self.container = [None] * self.size
        
    def prime_value(self):
        if self.size == 1:
            return "Not a prime value!"
        if self.size == 3:
            return 2
        if self.size == 4:
            return 3
        for i in range(self.size-1,4,-1):
            count = 0
            for j in range(2,i+1):
                if i % j == 0:
                    count += 1
            if count >= 2:
                continue        
            if count == 1:
                return i
        
    def get_hash_two(self,key):
        prime = self.prime_value()
        h = 0
        for char in key:
            h += ord(char)
        return prime - (h%prime)
        
    def get_hash_one(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h%self.size
    
    def merge_hash(self,key,const):
        h1 = self.get_hash_one(key)
        h2 = self.get_hash_two(key)
        h_val = h1 + (const * h2)
        return h_val%self.size
    
    def __setitem__(self,key,value):
        if not self.is_full():
            h = self.get_hash_one(key)
            if self.container[h] is None:
                self.container[h] = (key,value)
                return
            
            if self.container[h][0] == key:
                self.container[h] = (key,value)
                return
            
            for i in range(1,len(self.container)+1):
                val = self.merge_hash(key,i)
                if self.container[val] is None:
                    self.container[val] = (key,value)
                    return
        else:
            self.is_full()
    
    def __getitem__(self,key):
        if not self.is_empty():
            h = self.get_hash_one(key)
            if self.container[h] is not None:
                if self.container[h][0] == key:
                    return (h,self.container[h])

            for i in range(1,len(self.container)+1):
                val = self.merge_hash(key,i)
                if self.container[val]:
                    if self.container[val][0] == key:
                        return (val,self.container[val])
            return "Value not found!"
        else:
            self.is_empty()
            
    def search(self,key):
        if not self.is_empty():
            h = self.get_hash_one(key)
            if self.container[h] is not None:
                if self.container[h][0] == key:
                    return True
               
            for i in range(1,len(self.container)+1):
                val = self.merge_hash(key,i)
                if self.container[val]:
                    if self.container[val][0] == key:
                        return True
            return False
        else:
            self.is_empty()
    
    def __delitem__(self,key):
        if not self.is_empty():
            h = self.get_hash_one(key)
            if self.container[h] is not None:
                if self.container[h][0] == key:
                    val = self.container[h]
                    self.container[h] = None
                    return val

            for i in range(1,len(self.container)+1):
                val = self.merge_hash(key,i)
                if self.container[val]:
                    if self.container[val][0] == key:
                        value = self.container[val]
                        self.container[val] = None
                        return value
            print("Value not found!")
        else:
            self.is_empty()
    
    def get_count(self):
        count = 0
        for i in self.container:
            if i is not None:
                count += 1
        return count
    
    def is_empty(self):
        if self.get_count() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_count() == self.size:
            return True
        return False

if __name__ == "__main__":
    h = Hash(10)
    print(h.container)
    print(h.get_hash_one('march 6'))
    print(h.get_hash_one('march 15'))
    print(h.get_hash_one('march 7'))
    print(h.get_count())
    print(h.is_empty())
    h['march 7'] = 15
    h['march 6'] = 25
    h['march 15'] = 10
    print(h.get_count())
    print(h.container)
    print(h.get_hash_one('march 17'))
    print(h.get_hash_two('march 17'))
    print("---")
    print(h.get_hash_one('march 71'))
    print(h.get_hash_two('march 71'))
    print("---")
    print(h.get_hash_one('march 19'))
    print(h.get_hash_two('march 19'))
    print("---")
    h['march 17'] = 20
    h['march 71'] = 50
    h['march 19'] = 5
    print(h.container)
    print(h['march 6'])
    print(h['march 7'])
    print(h['march 17'])
    print(h['march 28'])
    print(h['march 71'])
    print(h.container)
    del h['march 17']
    del h['march 15']
    del h['march 71']
    del h['march 99']
    print(h.container)
    print(h.search('march 6'))
    print(h.search('march 28'))
    print(h.container)
    print(h.get_count())
    print(h.is_empty())
    print(h.is_full())
```

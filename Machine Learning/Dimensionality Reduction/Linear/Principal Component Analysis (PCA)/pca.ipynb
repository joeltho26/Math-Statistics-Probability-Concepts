#### Principal Component Analysis:
- here the PCA is used to perform dimensionality reduction thereby reducing the number of dimensions from the original dimensions by preserving their relationship between data points
- here the PCA focuses on principal component that contributes the highest amount variability of data

Process of PCA:
- the first step, we will project the data points in the 2 dimensional space
- now, we take the average of the data point values w.r.t x & y axes

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/pca.png')
new_image = img.resize((350, 260))
display(new_image)
```

- now, we take the intersecting point of the average values w.r.t x & y axes of the data points and moved to the origin

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/pca2.png')
new_image = img.resize((350, 160))
display(new_image)
```

- we will fit a best fit line now based on the datapoints that goes through the origin 
- to calculate the best fit line there 2 options
    - option 1: minimize Residual sum of squares
    - option 2: maximize the sum of squared distance of each data point from its origin
- we choose the second option since it alot simpler to perform the evaluation

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/pca3.png')
new_image = img.resize((350, 260))
display(new_image)
```

- to evaluate the best fit line based on the second option, we choose the data points and project it on the line fit and then the distance is calculated between the origin and the data points itself and the distance is squared

#### Linear Discriminant Analysis:
- it is a type of dimensionality reduction algorithm
- here while reducing the number of dimensions i.e. transforming to lower dimension by preserving the relationship between data points from the original higher dimension
- LDA focuses on the maximizing the separability among known categories or clusters.
- we will try to fit a best fit line by projecting data points on the line and calculate the separability based on mean and dispersion values of known clusters/categories

- calculate the mean value of each category/cluster and also calculate the value of spread i.e. dispersion w.r.t each category/cluster

- if there are 3 categories,
    - first we will calculate the centroids w.r.t all the data points
    - next, we will calculate the centroids w.r.t all data points associated to its category
    - now we will calculate the distance between each category centroid to its centroid of the overall data points and maximize it
    - we will also minimize the dispersion of data points w.r.t each know category
    - here since there are 3 categories, the LDA created 2 lines to separate the data
- LDA works similar to PCA with recards to the PC 1 giving the best variation of data and followed by PC2

Disadvantages:
- expects features to be gaussian distributed
- sensitive to outliers
- number of components is limited to the number of class labels
- suffers from small sample size problem or curse of dimensionality
- requries labelled data

#### Singular Value Decompostion:
- it is matrix factorization which is used to decompose a matrix into 3 simpler matrices

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/svd.png')
new_image = img.resize((150, 60))
display(new_image)
```

- the example for SVD is a table (matrix) consisting of people who rated movies
    - rows: people (U)
        - mxm orthogonal matrix whose columns are the left singular vectors of A
    - columns: ratings (Σ)
        - m×n diagonal matrix containing the singular values of A in descending order
    - cells: movies  (Vt)
        - n×n orthogonal transpose matrix whose columns are the right singular vectors of A

#### UMAP - Uniform Manifold Approximation & Projection:
- it is a non linear dimensionality reduction technique
- it is used to visualize the higher dimensional data points in lower dimension and understand its relationship
- while performing UMAP, transforms data from higher dimensions to lower dimensions, we preserve it relationship of data points that existed in the higher dimension

- now, relative to each data point, we will calculate the corresponding similarity scores by fitting a curve i.e. probability density function
- the curve fit will be different relative to each data point
- the probability density function i.e. the curve is produced with the help of high dimensional neighbours
- the default count of high dimensional neighbours is 15

- we will consider a data point "a" and look for the nearest data point based on the calculated similarity score

- the goal of this UMAP is to move data points closer to one another which are similar in high dimension and farther away from the data points which are not similar.
- to know the value of distance the data point to move closer and away, we fit a t distribution over the data point which is decided to move "b" which is also the similarity value for lower dimension of data points

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/umap14.png')
new_image = img.resize((350, 210))
display(new_image)

from PIL import Image 
from IPython.display import display

img = Image.open('images/umap18.png')
new_image = img.resize((150, 50))
display(new_image)
```

- the t distribution curve has flattened tails, the value of distance to move for the data point b is identified by projecting the data point onto the curve and chose the relevant value
- t-distribution curve is fit on "d" and "a" points

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/umap15.png')
new_image = img.resize((350, 120))
display(new_image)
```

- This process is continued iteratively, until the same distance structure in higher dimension w.r.t data points is formed in the lower dimension
- the advantage of UMAP is that, this process of moving data points in lower dimension closer to each one and farther away based on the similarity values is not just for 1 data points at each instance, rather UMAP  performs for few subset of data points simultaneously
- the high dimensional neighbour value is chosen carefully, since the lower neighbor value will focus on the independent cluster details and if the neighbour value is higher it focuses on the big picture
- UMAP uses stochastic gradient descent algorithm to reach global minimum in the convex curve of the cost function

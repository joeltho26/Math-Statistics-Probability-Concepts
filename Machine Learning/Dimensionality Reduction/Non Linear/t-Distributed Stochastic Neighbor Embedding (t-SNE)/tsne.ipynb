#### TSNE- T-Distributed Stochastic Neighbor Embedding:
- it used to perform dimensionality reduction in order to visualize the relationship between class data points in the higher dimensions
- its very similar to UMAP

- the first step of t-sne is that it projects the data points in the original higher dimensions and then we calculate the distance between the data points and these data points are translated into similarity scores
- We calculate the similarity score for each data point with relative to a data point by fitting a normal curve
- the point extended to the curve and that point on the curve gives the similarity score value

#### Self Organizing Maps:


- also called as Kohenon Maps
- it uses a Neural Network
- its works with the combination of neurons and input data
- first, consider the data points are in an N dimensional space along with the neurons i.e. weight units (2 units i.e. 2 clusters)
- at each iteration, we choose the 1 data point in the n dimensional space and compute the neuron
    - computing the neuron means updating the weight units by calculating the distances between input data and the weight units
    - now, among the 2 weight units we choose the weight unit closer in terms of the distance

Expectation Maximization:
- it is used to find the optimal parameters of the model, specifically the means, variances, and mixing coefficients, by iteratively maximizing the likelihood function
- there are 2 steps:
    - Expectation:
        - calculate the expected values of log likelihood function given the current parameters estimates
    - Maximization:
        - updates the parameter estimates to increase the expected log likelihood values

- the expectation step initial value is obtained by taking,
    - the means of each gaussian PDF from the clusters
    - the covariance
    - the weight

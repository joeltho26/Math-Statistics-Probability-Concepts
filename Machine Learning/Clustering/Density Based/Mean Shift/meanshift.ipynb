#### Meanshift Clustering:
- it is a type of clustering used for dense data points
- used to cluster non linear data points

- we set a window radius = R
    - R value is chosen based on kernel function
- at first, we take the R value and draw cirle on the dimensional space where the data points exist but anywhere
    - region of interest
- we, take the mean of the data point within the circle to identify the centre of mass
    - we find a shift/change in the mean value
        - mean shift vector

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/meanshift.png')
new_image = img.resize((350, 280))
display(new_image)
```

- the new mean value will be the centre for the next circle

#### Optics (Ordering Points To Identify the Clustering Structure) Clustering:
- type of density based clustering
- used when data is non linear
- when data is dense
- very similar to DBSCAN algorithm
- with the help of reachability distance, data points are clustered

- at first, we set the ε value and min_point value along with the core_distance value
    - ε
    - min_pts
    - core_dist

- just like DBSCAN algorithm, we draw the circle randomly anywhere on the dimensional space with datapoints
    - circle with radius R given by ε
- now check for the core point which satisfy the min point condition
    - min_pts = 5
- we also have another value called the core_dist = 3


```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/step1.png')
new_image = img.resize((420, 400))
display(new_image)
```

- once the min_point condition is satisfied, we look for the sub circle for radius is equal to core_dist value with the same center
- after drawing the circle, if we see again the min_pts condition satisfied
- then classify the new cluster as the circle with radius = 3mm

- now, we see the data points outside of the core distance are also core points of the first initial cluster and their core_dist value is undefined as shown below

#### Denclue (DENsity CLUstering) Clustering:
- type of density clustering

Advantages:
- small to medium datasets
- when data is dense and non linear
- when unknown cluster

Disadvantages:
- not used for large datasets and higher dimensions of data
- sensitive to outliers
- sparse data

- it inspired by kernel desnity estimation
    - estimate probability density function for an random variable
- at first, we fit the kernel density estimation function on top of each datapoint, so we see the curve is tall when the data is denser
- we have 5 kernel functions fit under the curve since there are 5 data points
    - the value under curve for each kernel density estimation = 1
    - for 5 data points, but we take the average to keep the area under the curve = 1
        - (5 * 1)/5 = 1

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/step1.png')
new_image = img.resize((500, 250))
display(new_image)
```

- for the data points closer, there is influence and result in a stronger and steeper curve
- the kernel density estimate when applied to each data point also uses a smoothing parameter to smoothen the probability curve
    - smoothing parameter = h
        - the higher h value the lower the curve
        - the lower h value the higher the curve

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/step2.png')
new_image = img.resize((400, 200))
display(new_image)
```

- if the h value is higher, we will have a smoothed out curve and we will include data points farther from it into the cluster too
- if the h value is lower, we will have a steep curve and we will exclude data points even which are closer and result in many outliers

- we have the density attractors which are the peaks which satisfy the threshold ε
    - data points closer to the density attractors will combine to form the clusters
    - density attractors are local maxima
- here the clustering can be partitional or hierarichal
    - hierarichal => based on the hyperplane
    - partitional => based on the density attractors

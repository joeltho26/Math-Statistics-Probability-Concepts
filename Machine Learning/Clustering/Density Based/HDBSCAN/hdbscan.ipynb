#### HDBSCAN:
- it is an extension of DBSCAN which overcome the issue in DBSCAN 
- it can handle clustering data points with varying density

Transforming the Space:
- transform the feature space using metric known as mutual reachability distance
    - calcuate the distance between 2 points
- by transforming the feature spacing with mutual reachability, helps us to emphasize the density based relationship between points

Constructing the Minimum Spanning Tree:
- once the mutual reachability is calculated between all data points, we construct a minimum spanning tree (MST) based on the distances between them
- the tree is connect all point in a way such that it minimizes the sum of edge weights
- Prim’s graph algorithm can be used to contruct the tree

Building the Hierarchy:
- constructs a hierarichal tree such as dendograms, 
- each leaf node represents the single data point and merge the data points based on the smallest edge weights i.e. mutual reachability distance

Condensing the Tree:
- now we condense the tree, based on the stability of cluster over varying density threshold
- clusters that are stable over the varying density threshold are choosen final and remaining are pruned


Extracting the Final Clusters:
- data points that do not belong to the stable clusters are denoted as noise points or outliers and denoted by -1
- can also provide soft clustering by providing the probability value for each data point belonging to the cluster

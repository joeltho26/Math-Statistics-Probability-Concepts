#### DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clustering:
- used to cluster data points that are dense in nature
- works great with non linear data
- challenged when working with high dimensional data
    - due to curse of dimensionality

- at first, we set the values for ε Epsilon and min_point
    - ε => radius of the circle with data point as the center
    - min_point => minimum number of points (threshold)

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/properties.png')
new_image = img.resize((200, 200))
display(new_image)
```

- there are 3 types of data points here
    - if min_pts = 3
    - core point
        - satisfy the min_point condition
    - border point
        - doesnt satisy min_point condition but there are data points existing within the circle greater than 1
    - noise point   
        - there is only 1 point within the circle 
        - not satisfying the min_point condition
- if we use min_pt as 1, then we will cluster even the noise point and the cluster of data point is not correct

- lets start with unclustered data

#### Rand Index:
- used to evaluate the similarity between clusters
- measures the percentage of correctly clustered data points
- value range 0 t0 1
    - higher the value, the better
- limiation: sometimes, the rand index value will be higher even if the cluster are formed properly and the agreement occuring by chance

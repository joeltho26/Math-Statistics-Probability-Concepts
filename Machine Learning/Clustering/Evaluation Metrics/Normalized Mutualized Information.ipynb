#### Normalized Mutual Information (NMI):
- used to evaluate the clustering of groups
- compared the ground truth to predicted clusters
- quantifies the amount of information that one clustering provide about the other.
- measure the amount of reduction in uncertainity about 1 cluster to the knowledge of the other
- 1.0: indicates perfect agreement between the cluster assignments and the ground truth labels. The clusters perfectly match the classes. Â  
- 0.0: indicates no mutual information between the cluster assignments and the ground truth labels. The clustering is independent of the true class structure.    
- H(Y) is the entropy of the true class labels (Y)
- H(C) is the entropy of the cluster assignments (C)
- I(Y; C) is the Mutual Information between (Y) and (C)

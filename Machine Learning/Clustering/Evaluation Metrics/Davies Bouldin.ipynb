#### Davies-Bouldin Index:
- measures the average "similarity" between each cluster and its similar cluster
- the similarities are the ratio of distance of data points between intra cluster to the inter cluster
- the lower the value, the better
- the DBI formula,
    - k => no. of clusters
    - Ri => similarity measure for cluster (i)

#### Adjusted Rand Index:
- measures the similarity between 2 clusters
- compares the ground truth and algorithm result
- it overcome the agreement by chance by the RI
- Adjusted Rand Index:
    - E[RI] => expected value of the Rand Index 
    - max[RI] => maximum value of RI = 1
    - ARI = 1 => perfect match
    - ARI = 0, agreement between the cluster formed occuring by random chance
    - ARI = -1, agreement between the cluster formed is worse than occuring by random chance

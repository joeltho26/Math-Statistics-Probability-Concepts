#### Homogeneity score:
- ranges: 0 - 1
- measure if the data points present in the cluster are from 1 class
- the higher, the better
- (H(C|K)) is the conditional entropy of the class labels (C) given the cluster assignments (K)
- (H(C)) is the entropy of the class labels (C)

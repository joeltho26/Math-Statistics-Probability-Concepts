#### Silhouette Score:
- used to idenfity the quality of cluster
- calculate the average intra cluster distances between the points
- calculates the average inter cluster distances between the point "i" from 1 cluster to the points in the neighbouring cluster
- we take evaluate the difference in above calculates distances
- we then divide the difference taking the max of either of the distances calculated    
- SS = +1 => indicates good clustering
- SS = 0 => indicates overlapping clusters
- SS = -1 => data points belonging to wrong clusters

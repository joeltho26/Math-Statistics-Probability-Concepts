#### Cophenetic Correlation Coefficient:
- measure how well dendogram preserves the pairwise distance or dissimilarities between the original points
- calculated for dendogram at different threshold
- the cophenetic distance is the distance associated with the vertical lines in the dendogram
- higher the value, the better
- to calcuate the cophenetic correlation coefficient
    - cophentic distance between data points
    - original distance between data points
        - euclidean, manhattan, minkowski, etc
    - average of original distances of all points
    - average of cophenetic distance of all points

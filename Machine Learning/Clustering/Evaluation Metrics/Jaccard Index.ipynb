#### Jaccard index:
- it compares two clustering solution
- it measures the ratio of correct clustered data points to the overall data points from a class and the falsely predicted data point into the cluster
    - solution 1: ground truth
    - solution 2: algorithm clustering
    - Jaccard = TP/TP+FP+FN
    - the higher the valuer, the better

#### Calinski-Harabasz Index:
- also known as Variance Ratio Criterion
- ratio of (sum of inter cluster dispersion)/(sum of intra cluster dispersion)
    - sum of inter cluster dispersion
        - B => Between-Cluster Sum of Squares
    - sum of intra cluster dispersion
        - W => Within-Cluster Sum of Squares
    - N => number of data points
    - k => number fo clusters
- higher the value, the better

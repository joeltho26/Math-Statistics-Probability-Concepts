#### Affinity Propogation:
- it is an unsupervised algorithm to classify data points into certain clusters based on their similarities
- the number of clusters need not be specified
- passes messages between pairs of data points to form clusters until convergence

- at first, calculate the similarity matrix for all pairs of data points 
- similarity:
    - s(i,k) quantifies how well suited the data point "k" to serve as the exemplar (centroid for a cluster) for data point "i"
    - similarity is calculated by -ve euclidean distance 
        - s(i,k)=−(∣∣xi −xk∣∣)^2
- preference:
    - influences the number of clusters
    - the diagonal values of similarity matrix
    - higher preference makes the data point more likely to become an exemplar (centroid)
    - initially the data points "i"
    - preference for all data points is intialized to median similarity value of the input matrix
- messages:
    - initialize the responsibility and availablity messages = 0
    - responsibility:
        - message sent from data point i to candidate exemplar k
        - there are rules associated to responsibility messages and are updated accordingly in each iteration until convergence

- examplar:    
    - after several iterations, the algorithm determines the exemplars
    - an exemplar is a data point k for which the sum of responsibility and availability equals a positive value

- cluster assignment:
    - each non exemplar data point "i" is assigned to the cluster whose exemplar k maximizes the sum of responsibility and availability scores/values

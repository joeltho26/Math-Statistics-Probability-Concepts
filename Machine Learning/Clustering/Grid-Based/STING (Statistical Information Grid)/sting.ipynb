#### StInG - Statistical Information Grid:
- it is a grid based algorithm
- used for spatial data mining

- the spatial data is divided into rectangular cells at different levels of resolution and the cells form a tree structure
- the cell size at higher level will be smaller and vice versa

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/ex.png')
new_image = img.resize((450, 270))
display(new_image)
```

- for each cell we calculate the below parameters
    - mean, count, min, max, SD, type of distribution
- calculation of these parameters starts from root to the bottom leaf nodes and check for the dense regions
- we calculate the confidence intervals of each cell at current level using the computed statistical information
- we label each cell as relevant or irrelevant based on confidence intervals
- but if the current level is not the bottommost level in the tree
    - then we descent to a lower level and repeat the above steps of calculating the confidence intervals and label them as relevant and non-relevant
- once the bottommost level reached, we look for the relevant cells and form them as clusters
- we can refine the cluster formed further by processing the data points on the relevant cells

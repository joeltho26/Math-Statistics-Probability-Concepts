#### CLustering In QUEst:
- it is a grid based algorithm
- it is used to identify cluster in subspace of the data which is common in high dimensional datasets

- Partitioning the Data Space:
    - for each dimension, divide the data range into predefined number of equal intervals
    - the above step creates a grid structure over the data space

- Identifying Dense Units in each 1 Dimension:
    - count the number of data points falling into each interval
    - if the fraction of data points exceeds the density threshold, mark the unit as dense
    - Eg: for Age & Salary:
        - we found the dense units as shown below

#### Cure Clustering:
- it is a type of hierarichal clustering
- when the number of clusters ar unknown
- it can handle high dimensions of data
- it has challenges when handling large dataset
- works for both linear and non linear datasets

- at first, we draw a random sample from large dataset
- partition the sample into small subsets
- perform agglomerative hierarichal clustering for the partitioned subsets until certain clusters are reached
- representative points:
    - now, each partial cluster contains not just one centroid but select C number of points as centroids that are well scattered points
- shrinking: 
    - after choosing the representative points, the points are shrunk towards the cluster centroids by fraction
    - by shrinking, we reduce the effect of outliers
- now, merge the partial cluster based on the minimum distance between the representative points and repeat until the certain number of clusters formed
- assign the remaining points to the cluster whose representative point is closer to them

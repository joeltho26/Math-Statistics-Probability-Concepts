#### Birch (Balanced Iterative Reducing and Clustering using Hierarchies):
- it an unsupervised data mining algorithm
- uses the calculation of Radius & Cluster Feature
    - Cluster Feature CF:
        - N = number of data points
        - LS = sum of the data points
        - SS = sum of squares of the data points
    - calculate Radius using CF properties
    - if R>T (threshold), we create a new CF or we append the data point to the existing CF
    - having more than 1 CF, we need to check the nearest CFs for the data points which needs to be added to the clusters

- lets consider the data points for clustering algorithm
- we also consider the threshold value to be set to T < 1.5 and Max Branch = 2 (branching factor)
- branching factors tells us about how many cluster feature we can have within 1 node
    - in other words, it says the maximum number of child nodes that each non leaf node in CF tree can have

- for each data point, we calculate the radius and the cluster feature
- now lets consider the first data point (3,4)
    - as there is only one entry in the feature map, the radius = 0
    - calculate the cluster feature
        - CF(N,LS,SS)
            - N = number of data points under consideration
                - N = 1
            - LS = Sum of data points under consideration
                - LS = (3,4)
            - SS = Square sum of data points under consideration
                - SS = (3^2,4^2) = (9,16)
        - CF value = (1,(3,4),(9,16))
    - now create a tree with CF value as root and leaf as data point (3,4)

#### Divisive - Hierarichal clustering:
- at first, the clustering algorithms begins with clustering all the data points into 1 single large cluster
- now the algorithm, will try to split the cluster into 2 or more clusters using the K-means flat clustering algorithm
- the cluster to split further down is chosen based on the below options
    - cluster with biggest diameter
    - cluster with the highest SSE
    - cluster with dissimilar points
- we continue the split of cluster until, we get each data point as an individual cluster
- we simultaeously, create the dendograms for the process of splitting the cluster
- we have a threshold value, for which we will cut the dendograms horizontally
- the number of vertical lines for which the threshold value cut is chosen as the optimal number of clusters
- to choose the optimal value for threshold, we can use several cluster evaluation metrics

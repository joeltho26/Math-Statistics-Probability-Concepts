#### Chameleon Clusteing:
- it is a heirarichal type of clustering
- it works on the graph principles

- from the dataset, we construct the proximity matrix or similarity matrix
    - these are the weights or similarities of the edges connecting two nodes
- the proximity matrix is then used to build a sparse graph
    - using KNN to find the nearest neighbor and connect the data points like nodes in a graph
    - we need 3 or more edges from a node in graph generally

- then, we partition the sparse graph into smaller cluster which are tightly coupled by cutting the edges
    - we use the threshold "min_size" to perform partitioning
        - we cut the graph into 2 if there are mini clusters greater than the min_size
        - we perform this iteratively

- the partitioned graph is again merged to form clusters based on
    - relative interconnectivity
        - greater the value the better
    - the below image,
        - C1 (left), C2 (right), C1, C2 merged near the red line using the nodes and their weights

- EC(Ci,Cj) = the sum of edges that connect clusters C1 & C2
- EC(Ci) = the minimum sum of the cut if we bisect C1
- EC(Cj) = the minimum sum of the cut if we bisect C2

- relative closeness:
    - looks for the merged clusters to resemble very close to each sub clusters
        - Mi,Mj = size of the clusters
        - Sec(Ci,Cj) = average weights of the edges that connect both the clusters
        - Sec(Ci) = average weights of the edges in cluster Ci
        - Sec(Cj) = average weights of the edges in cluster Cj
    - greater the relative closeness value, better it is

#### Agglomerative Approach:
- it is a type of hierarichal clustering
- it is bottom to top approach
- uses dendograms to identify the number of clusters

Advantages:
- small to medium datasets
- when data has a hierarichal structure
- when unknown cluster

Disadvantages:
- not used for large datasets
- sensitive to outliers
- sparse data
- difficult handling different data types of data

- at first, we consider each data point as a cluster
- then, we join the nearest data points and form that a new cluster
    - the nearest points are calculated via distance metric
        - euclidean
        - manhattan
        - minkowski
        - jaccard
        - cosine
        - etc
- we repeat the step of combining the sub clusters to form larger cluster
    - to join 2 clusters to form a large cluster we can calculate the distances again
        - nearest points or single linkage
        - centroids of clusters
        - farthest point in each cluster or complete linkage
        - average of clusters or average linkage
        - ward:
            - minimize the total within cluster variance
    - each of the linkage techniques has a cost value (Î´) associated to it with a formula

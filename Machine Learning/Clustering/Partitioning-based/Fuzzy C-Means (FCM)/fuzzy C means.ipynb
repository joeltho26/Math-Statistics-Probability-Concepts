- Soft Clustering:
    - one data point is associated into a cluster with a certain probability known as membership value
    - Fuzzy C-means clustering

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/soft.png')
new_image = img.resize((290, 250))
display(new_image)
```

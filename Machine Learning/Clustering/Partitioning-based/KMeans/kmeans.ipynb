#### K Means Clustering:
- it is a type of partitioning clustering
- uses the mean values to calculate the centroid of each cluster

Advantages:
- when the cluster value is known
- when clusters are non overlapping
- works good for large datasets
- when clusters are expected to be spherical

- K Means uses the LLoyd's algorithm
- the first step of the K-Means is to assign the values for 
    - K (no. of cluster) and their centroids
- initially, we choose random data points on the graph as the centroids to compute the cluster.
- however, sometimes, we do not get the optimal centroids of the cluster and clusters are formed incorrectly

- to form the correct centroids for K cluster, we use K-Means++ technique.

K-Means++:
- first we take a random point p1
- now we take a point p2 that is farthest from the point p1 and mark it as centroid C1
    - we calculate the distances & normalize it

There are 2 important metrics in K-Means:
- Distortion:
    - measures the avergae distance between the datapoints and the centroid of the cluster
    - lower distortion value indicated better clustering

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/distortion.png')
new_image = img.resize((400, 60))
display(new_image)
```

- Inertia:
    - sum of squared distances of each data point to its closest cluster center
    - Inertia is the numerator of the Distortion

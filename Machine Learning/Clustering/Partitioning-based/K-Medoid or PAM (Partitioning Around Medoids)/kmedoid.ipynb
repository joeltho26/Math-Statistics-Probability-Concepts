#### K-Medoid Clustering Algorithms:
- it is a type of partitioning algorithm
- used to cluster data points into groups based on similarities
- it is very similar to K-Means but instead of means, here we calculate the medoids values
- K medoid algorithms is also known as Partitioning Around Mean

Advantages:
- less sensitive to outliers
- easy to implement
- faster to converge compared to K-Means

Disadvantage:
- not ideal for clustering non spherical group of objects

- at first, we choose K value (no. of clusters)
- similary to K-means, we choose the optimal K-value and its centroids
    - Cross Validation and Hyperparameter tuning
    - Elbow method
        - WCSS or inertia
    - for optimal centroid
        - use Kmeans++

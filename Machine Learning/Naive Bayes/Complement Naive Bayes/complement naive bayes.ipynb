#### Complement Naive Bayes:
- it is the same as multinominal naive bayes algorithm
- it is used to handle imbalanced datasets
- instead of calculating the probability of item belonging to a class, we calculate the probability of the given item not belonging to it

- consider the dataset below for the example for complement naive bayes

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/data.png')
new_image = img.resize((590, 250))
display(new_image)
```

- just as multinomial naive bayes, we calculate the prior probabilities
    - P(Apples) = 2/3
    - P(Bananas) = 1/3
- calculate the count of labels associated with output labels
    - Apples = 8
    - Bananas = 16
- now we have a test data to be classified to a certain output class label

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/test.png')
new_image = img.resize((500, 150))
display(new_image)
```

- to classify the test data for Apples we associate the denominator with the opposite class, 
    - P(Round=1,Red=1,Soft=1|Apples) = 2/3 * (1/((1/16)*(5/16)*(1/16))) = 6.302
    - P(Round=1,Red=1,Soft=1|Apples) = 1/3 * (1/((1/8)*(1/8)*(2/8))) = 85.333
- the output label is predicted based on the value being the smallest compared the both 
    - output label = Apples

#### Bayes Theorem:
- used in ```bayesian inference statistics```
- also used in various machine learning algorithms
- evaluates the ```conditional probability``` of an event B given the event A has already occured
- calculates the ```posterior probability``` given,
    - ```likelihood```
    - ```prior probability```
    - ```marginal probability```

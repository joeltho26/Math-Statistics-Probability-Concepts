#### Gaussian Naive Bayes:
- used when input data is ```continuous numerical feature```
- used to solve classification problems
- works on the principle of ```bayes theorem```
- ```prior probability``` and ```posterior probability``` values are calculated the same as multi-nomial naive bayes
- to calculate ```likelihood value``` instead of using probability of frequency/count of words, it is obtained from ```distribution curve``` w.r.t data points that belong to particular output class and feature
- after calculating likelihood value, we take ```log``` of them because sometimes the likelihood is very tiny number which is difficult to compute
- final output class chosen based on ```highest posterior probabilty``` value produced among the classes

- the likelihood values for each feature is evaluated using a line projected onto the distribution curve
- calculate ```posterior probability``` using ```Likelihood``` values of Class ```"loves Troll 2"```:
    
    - ```P(Popcorn=20,Soda Pop=500,Candy=25|Loves Troll 2)```
        
        ```= P(Popcorn|Loves Troll 2) * P(Soda Pop|Loves Troll 2) * P(Candy|Loves Troll 2) * P(Loves Troll 2)```
        
        ```= 0.06 * 0.004 * really small number```

        - when we have really small number, we will experience ```underflow```
            - machines will struggle to keep up the lower values after a certain decimal points
        - to overcome underflow, we perform taking natural log (ln or log base e) of the values
        - log (0.06) * log(0.004) * log(really small number) * log(0.5) =  - 2.8 - 5.5 - 115 - 0.69 = -124

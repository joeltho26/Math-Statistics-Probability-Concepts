#### Complement Naive Bayes:
- it used on ```imbalanced datasets```
- the ```prior probability``` value is calculated as same as any other naive bayes algorithm
- to calculate the posterior probability instead of considering the ```likelihood``` values that belonging to the respective output class, we consider the values that ```do not belonging to it```
- final chosen output class based on the ```lowest posterior probablility``` value

```Working of Complement Naive Bayes```:
- the sample imbalanced dataset below

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/data.png')
new_image = img.resize((460, 200))
display(new_image)
````

- we calculate the ```prior probabilities```

    - ```P(Apples) = 2/3```

    - ```P(Bananas) = 1/3```

- calculate the ```count of labels``` associated with output labels
  
    - ```Apples = 8```
    
    - ```Bananas = 16```

- a test data is shown below

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/test.png')
new_image = img.resize((320, 100))
display(new_image)
````

- to classify the test data for ```Apples/Bananas``` we associate the likelihood values of opposite class, 
    
    - ```P(Round=1,Red=1,Soft=1|Apples)```
    
        ```= 2/3 * (1/((1/16) * (5/16) * (1/16)))```
        
        ```= 6.302```
    
    - ```P(Round=1,Red=1,Soft=1|Bananas)```
    
        ```= 1/3 * (1/((1/8) * (1/8) * (2/8)))```
        
        ```= 85.333```

- the output label prediction is based on the ```smallest``` value of posterior probability
    
    - ```output label = Apples```

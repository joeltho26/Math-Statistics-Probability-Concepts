##### Process of C4.5:
- at first, we calculate the overall entropy based on the output classes using the below image
    - overall entropy = Entropy Info(Dataset) = -(7/10) * log2 (7/10) -(3/10) * log2 (3/10) => 0.8807
- Now we calculate the Entropy Info, Split Info, Information Gain, Gain Ratio
    - For CGPA: >=9, >=8, <8
        - Calculate entropy Info,
            - Entropy(>=9) = -(3/4) * log2 (3/4) -(1/4) * log2 (1/4) => 0.8108
            - Entropy(>=8) = -(4/4) * log2 (4/4) -(0/4) * log2 (0/4) => 0
            - Entropy(<8) = -(0/2) * log2 (0/2) -(2/2) * log2 (2/2) => 0
            - Entropy Info(CGPA) = (4/10) * Entropy(>=9) + (4/10) * Entropy(>=8) + (2/10) * Entropy(<8) => 0.3243
        - Calculate the Information Gain: 
            - Information Gain = Entropy Info(Dataset) - Entropy Info(CGPA) = 0.8807 - 0.3243 => 0.5564
        - Calculate Split Info:
            - Split Info = (4/10) * log2 (4/10) + (4/10) * log2 (4/10) + (2/10) * log2 (2/10) = 0.5285 + 0.5285 + 0.4641 => 1.5211
        - Calculate the Gain Ratio,
            - Gain Ratio = (Information Gain / Split Info) => 0.5564/1.5211 => 0.3658

#### Classification - CART:
- its used for both regression and classification prediction
- it can work with both numerical and categorical features i.e. discrete and continuous values
- it can handle missing values
- it can handle pruning of trees
- feature importance

##### Process of CART - Classification:
- calculate the average weighted gini index value for every feature with relevance to the output labels
    - For Loves Popcorn:
        - Gini Index(Yes) = 1 - (1/4)^2 - (3/4)^2 = 1 - 0.0625 => 0.375
        - Gini Index(No) = 1 - (2/3)^2 - (1/3)^2 => 1 - 0.556 => 0.444
        - Weighted Average Gini Index = (4/7) * 0.375 + (3/7) * 0.444 => 0.405 
    - For Loves Soda:
        - Gini Index(Yes) = 1 - (3/4)^2 - (1/4)^2 = 1 - 0.0625 => 0.375
        - Gini Index(No) = 1 - (0/3)^2 - (3/3)^2 => 1 - 1 => 0
        - Weighted Average Gini Index = (4/7) * 0.375 + (3/7) * 0 => 0.214

- For Age:
    - For 9.5:
        - Gini Index(<9.5) = 1 - (0/1)^2 - (1/1)^2 = 1 - 1 => 0
        - Gini Index(>=9.5) = 1 - (3/6)^2 - (3/6)^2 => 1 - 0.5 => 0.5
        - Weighted Average Gini Index = (1/7) * 0 + (6/7) * 0.5 => 0.429
    - For 15:
        - Gini Index(<15) = 1 - (0/2)^2 - (2/2)^2 = 1 - 1 => 0
        - Gini Index(>=15) = 1 - (3/5)^2 - (2/5)^2 => 1 - 0.52 => 0.48
        - Weighted Average Gini Index = (2/7) * 0 + (5/7) * 0.5 => 0.343
    - For 26.5:
        - Gini Index(<26.5) = 1 - (1/3)^2 - (2/3)^2 = 1 - 0.554 => 0.446
        - Gini Index(>=26.5) = 1 - (2/4)^2 - (2/4)^2 => 1 - 0.5 => 0.5
        - Weighted Average Gini Index = (3/7) * 0.446 + (4/7) * 0.5 => 0.476
    - For 36.5:
        - Gini Index(<36.5) = 1 - (2/4)^2 - (2/4)^2 = 1 - 0.5 => 0.5
        - Gini Index(>=36.5) = 1 - (1/3)^2 - (2/3)^2 => 1 - 0.554 => 0.446
        - Weighted Average Gini Index = (4/7) * 0.5 + (3/7) * 0.446 => 0.476
    - For 44:
        - Gini Index(<44) = 1 - (3/5)^2 - (2/5)^2 = 1 - 0.52 => 0.48
        - Gini Index(>=44) = 1 - (2/2)^2 - (0/2)^2 => 1 - 0.5 => 0
        - Weighted Average Gini Index = (5/7) * 0 + (2/7) * 0.5 => 0.343
    - For 66.5:
        - Gini Index(<66.5) = 1 - (3/6)^2 - (3/6)^2 = 1 - 0.5 => 0.5
        - Gini Index(>=66.5) = 1 - (0/1)^2 - (1/1)^2 => 1 - 1 => 0
        - Weighted Average Gini Index = (6/7) * 0 + (1/7) * 0.5 => 0.429

- We have to pick the average value with the lowest gini index
    - since we have two values with the same lowest value, we can choose any one i.e. 15 or 44
- now compare the gini index for the 3 features in the dataset and choose the feature with the lowest gini index score as the root node

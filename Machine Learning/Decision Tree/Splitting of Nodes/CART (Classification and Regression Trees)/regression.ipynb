#### Regression - CART:
- its used for both regression and classification prediction
- it can work with both numerical and categorical features i.e. discrete and continuous values
- it can handle missing values
- it can handle pruning of trees
- feature importance

##### Process of CART - Regression:
- Take the Dosage feature
    - Calculate the Residual Sum of Squares for the first record vs the rest in the dataset
        - RSS (1 datapoint) => 0
        - RSS (rest) => 27468.5

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss1.png')
new_image = img.resize((580, 300))
display(new_image)
```

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss2.png')
new_image = img.resize((700, 300))
display(new_image)
```

- Next we calculate the Residual Sum of Squares of first & second record vs the remaining datapoints in the dataset
- Likewise, we calculate the Residual Sum of Squares for records incrementally vs the remaining data points in the dataset
    - usually the minimum number of samples required in each node is 20 but could be modified
- We now the choose the the record at which the Residual Sum of Squares was having the minimum value as the split condition i.e. RSS(Dosage) = 19564

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss_graph.png')
new_image = img.resize((480, 220))
display(new_image)
```

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss3.png')
new_image = img.resize((700, 300))
display(new_image)
```

- To find the regression value as output, take the average of the datapoints within the split condition
- now we go to the next feature
- Take the Age feature:
    - Calculate the RSS value for each data point incrementally and find the record which has the lowest RSS value for condition split i.e. RSS(Age) = 12017

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss5.png')
new_image = img.resize((460, 300))
display(new_image)
```

- now we go to the next feature
- Take the Sex feature:
    - there are 2 categories (Male/Female)
    - Calculate the RSS value for Male and Female categories and find the record which has the lowest RSS value for condition split i.e. RSS(Sex) = 20738 (Female)

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/rss6.png')
new_image = img.resize((460, 320))
display(new_image)
```

- now, we compare the RSS for all the features and choose the feature with the lowest RSS score as the root node i.e. RSS(Age)
    - RSS(Age) = 12017
    - RSS(Dosage) = 19564
    - RSS(Sex) = 20738

- Likewise, we can repeat it for the remaining features in the dataset and we can re-use the feature for splitting of node until the number of samples in the node is unsatisfactory
- Effective value is calculated by taking the average of the output labels (continuous value) present within the split condition of the independent variable

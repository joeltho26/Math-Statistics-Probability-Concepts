#### ID3 Algorithm:
Link: https://www.youtube.com/watch?v=coOTEc-0OGw&ab_channel=MaheshHuddar

- difficult handling continuous values
- tends to favor large number of distinct values
- lack of pruning
- sensitive to noise data and leads to overfitting
- does not handle missing values
- handles only classification problem

##### Process of ID3:
- calculate the Information Gain for every feature
    - first calculate the overall entropy of the dataset based on the dependent variables output classes using the below formula or from the image below
        - overall entropy = -(9/14) * log2 (9/14) -(5/14) * log2 (5/14) => 0.94

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/entropy.png')
new_image = img.resize((470, 130))
display(new_image)
```

- now calculate the entropy for the first feature  
    - eg: Outlook = Sunny, Overcast, Rain:
        - calculate the entropy for all distinct values in the feature 
            - Entropy(Sunny|Play Tennis) = -(2/5) * log2 (2/5) - (3/5) * log2 (3/5) => 0.971
            - Entropy(Overcast|Play Tennis) = -(0/4) * log2 (0/4) -(4/4) * log2 (4/4) => 0
            - Entropy(Rain|Play Tennis) = -(3/5) * log2 (3/5) -(2/5) * log2 (2/5) => 0.971
        - now we calculate the information gain from the below value
            - Gain(Outlook) => 0.2464

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/informationgain1.png')
new_image = img.resize((670, 180))
display(new_image)
```

- likewise, we repeat the same steps for each feature and calculate the entropy & information gain
    - eg: Temperature = Cold, Hot, Mild:
        - calculate the entropy for all distinct values in the feature 
            - Entropy(Cold|Play Tennis) = -(3/4) * log2 (3/4) - (1/4) * log2 (1/4) => 0.8113
            - Entropy(Mild|Play Tennis) = -(4/6) * log2 (4/6) -(2/6) * log2 (2/6) => 0.9183
            - Entropy(Hot|Play Tennis) = -(2/4) * log2 (2/4) -(2/4) * log2 (2/4) => 1
        - now we calculate the information gain below image
            - Gain(Temperature) => 0.0289

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/sunny.png')
new_image = img.resize((320, 180))
display(new_image)
```

- the root node is formed with a feature and their decision is in relevance with the output classes are branched outward
- now we first choose the sunny branch with relevance to the output classes
    - now we continue the same process from steps 1 till the step before.
        - overall entropy(Sunny) = -(2/5) * log2 (2/5) -(3/5) * log2 (3/5) => 0.97
        - Temperature: Hot, Mild, Cold
            - Entropy(Hot|Sunny of Play Tennis) = -(0/2) * log2 (0/2) - (2/2) * log2 (2/2) => 0
            - Entropy(Mild|Sunny of Play Tennis) = -(1/2) * log2 (1/2) -(1/2) * log2 (1/2) => 1
            - Entropy(Cold|Sunny of Play Tennis) = -(1/1) * log2 (1/1) -(0/1) * log2 (0/1) => 0
            - Information Gain(Temperature) =  0.570
        - Humidity: High, Normal
            - Entropy(High|Sunny of Play Tennis) = -(0/3) * log2 (0/3) - (3/3) * log2 (3/3) => 0
            - Entropy(Normal|Sunny of Play Tennis) = -(2/2) * log2 (2/2) -(0/2) * log2 (0/2) => 0
            - Information Gain(Temperature) =  0.97
        - Wind: Strong, Weak
            - Entropy(Strong|Sunny of Play Tennis) = -(1/2) * log2 (1/2) - (1/2) * log2 (1/2) => 1
            - Entropy(Weak|Sunny of Play Tennis) = -(1/3) * log2 (1/3) -(1/3) * log2 (1/3) => 0.9183
            - Information Gain(Temperature) =  0.0192
- now compare the information gain for the features and choose the feature with highest information gain

#### Missing Values in Decision Tree:

Categorical Variables:
- Either, we can fill the missing values with the most common value in the feature and replace the null value with it
- Or else, we can find for the correlated feature and assign the related value to the missing values in the feature

Numerical Variable:
- Either, we can fill the missing values with the mean or median value in the feature and replace the null value with it
- Or else, we can find for the correlated feature and use linear regression line to find the appropriate value to replace for the missing values in the feature

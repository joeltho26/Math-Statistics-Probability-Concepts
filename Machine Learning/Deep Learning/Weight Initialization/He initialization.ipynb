#### He Weight Initialization:
- it is also called as Kaiming initialization
- it was built to overcome the issue of vanishing gradient while using Xavier/Glorot with relu activation function 
- there are 2 types of distribution part of he weight initialization technique
    - normal and uniform 
- the weights for neural network are chosen between the ranges shown below for normal and uniform distribution

#### Xavier-Glorot Initialization:
- there are 2 types of distribution part of xavier/glorat weight initialization technique
    - normal and uniform 
- the weights for neural network are chosen between the ranges shown below for normal and uniform distribution 
- xavier glorot can be used with Sigmoid and tanh activation functions, but for relu it might lead to vanishing gradient scenarios

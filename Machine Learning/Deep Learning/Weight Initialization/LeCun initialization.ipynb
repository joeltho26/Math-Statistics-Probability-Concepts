#### LeCun Weight Initialization:
- there are 2 types of distribution part of lecun weight initialization technique
    - normal and uniform 
- the weights for neural network are chosen between the ranges shown below for normal and uniform distribution 
- lecun can be used with SELU, sigmoid and tanh activation functions, but for relu it might lead to vanishing gradient scenarios

#### Generative Adversarial Network:
- here the gan has two components 
    - dicriminator
    - generator
- Discriminator:
    - at first, we feed real data and the initially generated fake data from generator to the discriminator model to train and identify real and fake data properly
    - we have also initialized weights to the discriminator
    - while training the discriminator model, we have discriminator loss associated to it
    - at each iteration, we update the weights associated to the discriminator to identify and distinguish the images as fake and real more accurately
- Generator:
    - while training the discriminator, we also train the generator as well
    - the goal of generator is to create synthetic fake data which confuses the discriminator to identify which data is fake and real
    - at first we give the noise data into the generator to generate fake data 
    - now, we will take the generated noise data and give it to discriminator to identify the sample as fake or real and there is a realness score associated to it.
    - based on the score, we also have a generator loss value associated which will be improved at each iteration by updating the weights in the generator model. 
    - after several iteration, we would have built a generator model which will output fake data which the discriminator will find it hard to distinguish between real and fake data
    - now, we will drop the discriminator model and only keep the generator model to generate images

#### Parametric Relu Function:
- the range of parametric relu function is -negative values to positive values
- the derivative of parametric relu function is approximately between 0-1 but the values are greater than 0 (small values) to 1
- it helps to overcome dying relu or dead activation problem
- it overcomes vanishing gradient problem
- we have a lerning rate denoted by α
    - based on the α value, we transition the parametric relu to leaky relu or regular relu, etc
- the values are not zero centered

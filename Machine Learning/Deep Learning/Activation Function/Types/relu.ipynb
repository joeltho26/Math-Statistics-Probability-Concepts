#### Relu Function:
- it is called as rectified linear unit
- the values for relu ranges between 0 to positive value
- the derivative values for relu ranges between 0 to 1
    - any value greater than 0 then its 1 and if any value lesser than 0 then it is 0
- it has thew dying relu or dead activation function problem
- values are not zero centered
- it overcomes vanishing gradient problem

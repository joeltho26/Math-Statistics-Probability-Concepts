#### Swish Function:
- it used when there is a deep neural network model with 40 or more hidden layers
- it is not zero centered
- it is a self gated function
    - a self gated function only needs a simple single scalar input whereas the gated function needs a multi scalar input
    - it becomes easier to transform the activation from sigmoid scalar or relu activation to swish activation function in the hidden layers
- the range of swish function is from small negative value, zero value to the positive values
- the approximate value of derivative swish function is 0-1 but there are small negative value associated to it just like the regular swish function
- they are used in  Inception-ResNet-v2 and MobileNet hidden layers in CNNs, RNNs, GANs

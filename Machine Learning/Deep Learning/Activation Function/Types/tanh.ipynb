#### Tanh Function:
- the function value ranges between -1 to +1
- the tanh values are zero centered
- the derivative of tanh results in values ranging between 0-1
- tanh results in vanishing gradient problem 
- this activation function is used in RNN, LSTM, GRUs, GANs and implemented in the hidden layers of the neural network

#### Gelu Function:
- it is called gaussian error linear unit
- it is used to overcome dying relu problem
- it used in hidden layers of the neural network
- the values of gelu ranges between small negative value to positive values
- the values of the derivative of gelu function ranges between small negative to positive value
- it is used in transformer models, reinforcement learning LLMs, Computer vision, etc

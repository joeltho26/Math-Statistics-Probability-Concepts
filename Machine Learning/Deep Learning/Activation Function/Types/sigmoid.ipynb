#### Sigmoid Function:
- the range of the sigmoid function is between 0-1
- the derivative of sigmoid functon is between 0-0.25
- the sigmoid function is applied to the output layer when we need to predict binary class output
- the sigmoid function produces vanishing gradient problem
- the values of signmoid are not zero centered

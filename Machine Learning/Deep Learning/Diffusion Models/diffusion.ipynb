#### Diffusion Models:
- here the diffusion models have two process
    - forward and backward diffusion
- forward diffusion:
    - we iteratively add gaussian noise to the input data making it a markovs chain process
        - markovs chain process:
            - it is a stochastic process, i.e. the probability of transitioning to the next state depends on the current state and not the previous state
    - the data is transformed into pure random noise data
- backward diffusion:
    - model learns to reverse the forward diffusion process to generate the input data without noise
    - during model training process, noisy data is given and used to predict the noise in the data.
    - after iteratively training the model to predict noise in the data, we can denoise the input data and produce pure input data without noise
- used in image generation, 3d modeling, video generation, image editing, etc
- there are two types of Diffusion models working:
    - score based
        - adds noise via stochastic differential equations
        - during reverse diffusion process, it tries to predict the score function and compares with the original score function witout noise 
    - denoising 
        - noise added via gaussian distribution function
        - tries to denoise the applied gaussian noise at each iteration of backward diffusion process

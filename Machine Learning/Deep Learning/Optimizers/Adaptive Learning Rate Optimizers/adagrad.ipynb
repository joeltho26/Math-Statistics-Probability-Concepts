#### AdaGrad:
- it uses different learning rates during the back progation to reduce the loss function value
    - the learning rate will reduce during each step iteratively
- it useful when there is sparse and dense feature together used 

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/adagrad.png')
new_image = img.resize((390, 250))
display(new_image)
```

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/adagrad1.png')
new_image = img.resize((270, 250))
display(new_image)
```

- we have "ε" a small value added along with "αt" in the modifying learning rate const because if "αt" = 0 then the whole gradient term will become 0 and now the old weight and new weight calculated will be the same

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/adagrad2.png')
new_image = img.resize((320, 250))
display(new_image)
```

- the initial learning rates will cause the point to move towards the global minima and slowly the changes are reduced and takes small steps

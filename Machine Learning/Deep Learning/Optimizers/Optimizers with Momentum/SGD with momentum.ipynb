#### SGD with momentum:
- it overcomes the issue of noise when mini batch SGD is used as shown in the below image
    - the noise will extended the time in reaching the global minima
- the noise can be reduced by using exponential moving average which has a velocity term associated with the optimization
    - (γ * v) + (η * g) 
        - γ = momentum coefficient that ranges between 0-1
        - v = velocity
        - η = learning rate that ranges between 0-1
        - g = derivative of loss with weights (gradients)
- we calculate the gradient, velocity in each step
- disadvantage of SGM with momentum, we use perform the gradient at current step and then apply momentum which sometimes will result in overshooting the global minimum especially in sharp valleys

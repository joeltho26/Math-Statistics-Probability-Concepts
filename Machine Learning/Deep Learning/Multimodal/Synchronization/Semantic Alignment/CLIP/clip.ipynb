1. [CLIP](#clip)

#### CLIP:
- it is used to produce the embeddings in which the images and the text are in the shared latent space
- it was developed by Open AI for multimodel AI models
- it uses dual encoder
    - one encoder for image
        - produces image embeddings
    - another one for text
        - produces text embeddings
- both the embeddings are combined which represent the relationship between text and images
- CLIP training process:
    - contrastive pre-training
        - CLIP is pretrained on 400 million text-image pairs
        - it created matching and unmatching pairs of embedding
        - CLIP computes the embeddings between image and text encoder using cosine similarity and optimizes it
    - create dataset classifier from label text
        - for each image, several descriptions are created 
        - the descriptions are fed into the text encoder to generate class specific embeddings
        - it uses contrastive loss function
            - it rewards for matching pairs and penalizes for incorrect pairs 
    - zero shot prediction
        - now the trained text encoder used as a zero-shot classifier

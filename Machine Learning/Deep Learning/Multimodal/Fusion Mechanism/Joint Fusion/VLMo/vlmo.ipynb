#### VLMo - Vision Language Model:
- the input for VLMo is the image-text pair
- the image-text pair is encoded into image, text and image-text representation
- Image representation:
    - split and reshaped into N patches called path embeddings
    - the patch embeddings also consist of I_CLS a learnable special token is prepended to the sequence 
- Text representation:
    - tokenize into subword units
        - subwords eg: "looked" => "look" + "ed"
    - start of sequence token + special boundary token are added to the text sequence 
- Image-Text representation:
    - concatenate the text and image representations

- now the created image and text representations are given to the multiway transformer to encode different modalities 
- FFN consisting of 2 linear transformations and 1 activation is used to encode 

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/vlmo.png')
new_image = img.resize((790, 720))
display(new_image)
```

- here in multiway transformer, we perform three pretrainings
    - Vision pre-training
        - pre-train vision FFN and multi head self attention by using the image representation
    - Language pre-training
        - pre-train language FFN by using the text representation
    - Vision Language pre-training
        - pre-train vision language FFN and multi head self attention by using the concatenated image-text representation

Matching Image-Text model:

Masked Language model

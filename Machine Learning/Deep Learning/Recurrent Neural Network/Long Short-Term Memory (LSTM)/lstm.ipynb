#### LSTM - Long Short Term Memory
- it performs better than vanilla RNN by remembering the context information by the use of memory cell
- it has 3 gates and a candidate memory
    - forget gate
        - makes the lstm to forget few information
    - input gate 
        - regulate the addition of new information
    - candidate memory
        - add information into the memory cell to form new cell state
    - output gate
        - gives out the information to the next time step of the lstm (updated hidden state) by concatinating information from both updated memory cell and the input information using sigmoid fn.
- the hidden state and memory cell are maintained separately
- Disadv: 
    - higher training times for LSTM, sometimes considered ad black box model i.e. challenging to interpret how they arrive at a particular decision
- Adv:
    - avoid vanishing and exploding gradient
    - handling variable length sequences
    - memory cell to remember context for longer sequences

#### GRU - Gater Recurrent Unit
- it is a simplified version of LSTM
- it can limit their ability to capture very complex relationships or long term dependencies but it trains faster than LSTM
- it consist of a reset gate and an update gate
- in gru, it has both hidden state and memory cell together

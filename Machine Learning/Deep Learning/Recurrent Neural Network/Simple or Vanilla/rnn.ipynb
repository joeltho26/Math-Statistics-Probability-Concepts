#### Recurrent Neural Network:
- consider a rnn is similar to ann being squashed on top of each other and also giving the output weights of each ann as input to the next ann model
- a simple RNN is also called as vanilla RNN
- the rnn at each step takes the previous weight + initialized weight + input value for training and an output value
- it is used to process sequential data by including a feedback loop
    - speech recognition, language translation, time series, etc
- vanishing and exploding gradient descents, loss of context while processing long sequences, slow training times

#### Batch Normalization:
- used after convolutional layer and generally before applying the activation function.
- it is used to accelerate and stabilize the neural network training process
- it handles internal covariate shift
    - internal covariate shift:
    - when parameters of NN is constantly updated, there is a change/shift in the distribution of activations
    - these shift causes the model training process to slow and harder since these shifts affect the subsequent layers to adapt to the changes

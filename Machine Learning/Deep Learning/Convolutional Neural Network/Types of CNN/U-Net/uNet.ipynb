#### UNet:
- it is used in latent diffusion models and image segmentation (object detection)
- UNet's also have skip connections
- it consister of 3 important feature
    - Encoder:
        - use small 3x3 filters to scan, find and extract features
        - applies relu to add non linearity to the network
            - non linear activation functions are used to learn complex data relationships
        - 2x2 max pooling to shrink image size thereby resulting in dimensionality reduction
    - Bottleneck Layer:
        - its the place where the important extracted feature are in latent space and compressed 
    - Decoder:
        - uses upsampling to increase image size and get back to original image size
        - combines information from encoder via skip connections
    - 1x1 convolution:
        - at the end we use 1x1 convolution to generate segmentation map

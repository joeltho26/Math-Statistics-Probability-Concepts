#### ResNet:
- it has adopted functionality from VGGNet
- it uses multiple residual block which are stacked and consist of skip connection within them 
- in residual block there are multiple 3x3 convolutions 
- skip connection:
    - during training and weight updation, it skips few layers and directly bypasses to the later layers in network
- it overcomes vanishing & exploding gradient problems in NN
- its also faster than VGG19
- uses Global Average pooling as the final layer to reduce the training parametes thereby reducing the spatial dimensions of the feature maps

- Highway Networks:
    - there is a similar approach called highway networks which also uses skip connection

    - these gates determine how much information gets passed through skip connection
    - it doesnt have better accuracy than ResNet

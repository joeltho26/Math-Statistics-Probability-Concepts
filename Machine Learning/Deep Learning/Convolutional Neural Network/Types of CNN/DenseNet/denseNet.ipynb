1. [DenseNet](#densenet)

#### DenseNet:
- eliminates vanishing gradient problem
- revolutionary in computer vision field
- it consist of dense block, transitional layers, growth rate
- dense block:
    - consist of several layers of convolutions + batch normalization + activation function
    - the input given to convolution layer is the concatenated output from preceding convolutions layers
    - each layer produces k feature maps (k = growth rate i.e. hyperparameter)
    - the greater the k value, the more information is extracted and used 
- transitional layer:
    - reducing the number of feature maps and downsampling the spatial dimensions of the feature maps
    - batch normalization
    - 1x1 convolutions
    - average pooling
- variants of DenseNet:
    - Dense 121
    - Dense 169
    - Dense 201
    - Dense 264
- risk of overfitting, high computational cost, implementation complexity, high memory consumption

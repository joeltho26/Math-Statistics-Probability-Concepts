#### VGGNet:
- called as visual geometry group 
- there are 2 variants in VGG
    - VGG16
        - 13 convolutions layers, 5 pooling layers and 3 fully connected layers
        - activation functions: relu
        - pooling: max pooling
        - sotfmax function
    - VGG19
        - 16 convolutions layers, 5 pooling layers and 3 fully connected layers
        - activation functions: relu
        - pooling: max pooling
        - sotfmax function
- optimizers:
    - SGD momentum
- the input image for vggnet uses 224x224 image size
- slower than the newer ResNet architecture
- causes vanishing gradient problem

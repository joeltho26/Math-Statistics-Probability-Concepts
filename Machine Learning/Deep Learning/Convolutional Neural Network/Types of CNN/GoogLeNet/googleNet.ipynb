#### GoogleNet:
- it is also called as inception models
- it uses 27 layer deep network (included 5 pooling layer)
- features of GoogleNet:
    - 1x1 convolutions:
        - used thereby reducing the number of parameters in the model for training
    - Inception Module:
        - consist of inception module which performs multiple convolutions operations (1x1,3x3,5x5) in parallel along with max pooling and concatenate their outputs 
        - it is used to capture features of the input in multiple scales
    - Auxiliary Classifiers:
        - they are used to overcome vanishing gradient problem
        - it uses 1x1 convolution, average pooling, fully connected layer (relu), dropout layer, fully connected softmax layer, auxillary loss which is added to main loss with weight 0.3 value to stabilize training and regularize the network
    - Gobal average pooling
        - used at the end since fully connected layer increases the number of parameters for training the model and by using global average pooling we reduce the number of parameters

- there are 4 variants of inception models
    - inception v1:
        - initially introduced the inception module
    - inception v2:
        - replaced 5x5 convolutions with 3x3 convolutions
        - 7x7 convolutions replaced with 1x7 * 7x1
        - 3x3 convolutions replaced with 3x1 * 1x3
    - inception v3:
        - RMSprop optimizer
        - batch normalization used in auxillary classifier
        - label smoothing added to loss function to prevent overfitting i.e. to regularize
    - inception v4:
        - number of inceptions modules increased and uniform with same number of filters being used
        - there are 3 types of inception modules names as A,B,C
- there are also inception-resnet v1 & v2 architectures
    - resnet connections are introduced 
    - pooling layer is replaced by residual connections in inception blocks
    - network made deeper by increasing the number of inception modules

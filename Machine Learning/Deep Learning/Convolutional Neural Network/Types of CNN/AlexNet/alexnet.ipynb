#### AlexNet:
- it is a simple CNN architecture
- its has 5 convolution layer, 3 pooling layers, 2 normalized layer, 2 fully connected layers and 1 softmax layer
- pooling layer:
    - max pooling
- activations:
    - relu
- batch size:
    - 128
- optimizers:
    - SGD momentum
- Disadv: 
    - high computational cost, and relatively slow inference time

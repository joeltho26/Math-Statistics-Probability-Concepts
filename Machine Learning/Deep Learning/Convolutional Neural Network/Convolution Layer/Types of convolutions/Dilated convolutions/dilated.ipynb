#### Dilated convolutions:
- here we introduce holes and therby expand the receptive field of the filter without increasing the no. of parameters
- it has a dilation rate
    - eg: dilation rate = 2
        - would mean that the filter skips every other pixel in the input, effectively expanding its receptive field

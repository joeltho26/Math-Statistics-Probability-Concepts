#### Depthwise Separable convolutions:
- here we break the kernel dimensions from 3x3 into 2 1x1 kernels
- the 1x1 kernels are applied one after the other to perform convolutions
- thereby these types of convolutions are useful in mobile net architecture and they perform convolutions faster

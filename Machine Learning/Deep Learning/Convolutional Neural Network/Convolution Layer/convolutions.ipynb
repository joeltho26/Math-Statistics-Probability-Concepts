#### Convolutions:
- it is used to convert the image represented in higher dimensions based on the color scale to lower scale by capturing the important details and features
- a black and white image will be represented as a 2x2x1 or 3x3x1 matrix, channel = 1, whereas a color image will have 3x3x3 dimensional matrix where we have 3 channels with color Red, Green, Blue
    - the input matrix size for color images in most cnn are 224x224
- after convolution layer, we also use the activation functions

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/image.png')
new_image = img.resize((550, 260))
display(new_image)
```

- as part of the convolution, we use Filters (feature detectors or sometimes called as kernels) to generate the Feature Map
- Kernels & Filters are used interchangeably but they are different and they are used to generate feature map
    - filters and input matrix perform elemental multiplication to generate the feature or activation map

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/kernels.png')
new_image = img.resize((370, 300))
display(new_image)
```

![SegmentLocal](images/kernels.gif "segment")

- Kernels:
    - they are 2x2 or 3x3 or 5x5 or 7x7, etc matrices used in filters
    - they help us to capture information about the image such as edge detection, emboss, blurring, sharpening, etc
    - here in 2D convolutions uses 2x2 matrix

- Filters:
    - they are kernels stacked together to generate the feature map
    - here in 2D convolutions uses 3x3 matrix
    - the filter size is evaluated by hw * k
        - k = total number input channel
        - hw = dimensions of the kernel (2x2, 3x3, 5x5, 7x7, etc)

#### Backward Propogation:
- it is a vital step in neural network which is used to reduce the loss function and make the model more efficient to predict the output values/labels
- using back propogation, we can update the weights associated in the ML 
- the backward propogation using the gradient descent approach i.e. optimizer to update the weights
- the backward propogation takes the derivative of the output values and the weights via chain rule to update the weights and bias

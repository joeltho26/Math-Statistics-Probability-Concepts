#### Kolmogorov Arnov Networks:
- it is a very new architecture in Deep learning
- it is very similar to regular ANN but instead of having fixed activation function in each nodes of the hidden layer, we use a learnable function at the edge which connects each input to each node of the hidden layer
- after using the edge function at each edge which connects the input with nodes of the hidden layers we sum the values of each edge function before given to the node of the hidden layer

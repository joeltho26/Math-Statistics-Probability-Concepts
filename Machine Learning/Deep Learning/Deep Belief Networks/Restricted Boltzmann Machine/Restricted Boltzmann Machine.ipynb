1. [Restricted Boltzmann Machine]

#### Restricted Boltzmann Machine:
- it is an unsupervised machine learning 
- it tries to learn the patterns and the probability distribution of the data 
- an RBM consist of 2 neurons
    - visible layer
    - hidden layer
- "restricted" means there are no connection between the nodes of same layer i.e. hidden-hidden or visible-visible but there is a connection between each node in visible and hidden layers
- RBM's learn based on contrastive divergence algorithm:
    - using weights and bias, RBM try to maximize the probablilty of training data 
    - 3 important features:
        - positive phase
            - visible units propogate the activations to hidden units
        - negative phase
            - hidden units propogate the activations to visible units
        - weight adjustment 
            - weights are adjusted to increase the training data probability and decrease the probability of reconstructed data

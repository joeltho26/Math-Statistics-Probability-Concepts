#### Logistic Regression:
- it is part of the generalized linear models
- used for both classification and regression problem statements
- uses max likelihood to fit the best curve
    - calculate the likelihood for each data point and take the product of all
    - the max value of the product of all is chosen as the best fit

Note:
- we cannot use linear regression for classification problem because
    - the effect of outliers will affect the line fit 
    - if the values are higher and goes beneath 0 or beyond 1

- lets consider a binary classification problem for logistic regression

- we use to the logit function to convert the probabilities into log(odds) values i.e. convert probabilities between 0-1 to -infinity to 
+infinity
- for p=0.5, log(odds) = log(p/1-p) = log(0.5/1-0.5) = log (0.5/0.5) = log(1) = 0 
- for p = 0.731, log(odds) = log(p/1-p) = log(0.731/1-0.731) = log (0.731/0.269) = log(2.717) = 1
- for p = 0.88, log(odds) = log(p/1-p) = log(0.88/1-0.88) = log (0.88/0.12) = log(7.33) = 2
- for p = 0.95, log(odds) = log(p/1-p) = log(0.95/1-0.95) = log (0.95/0.05) = log(19) = 3

Notes:
- the logit function converts probability into log(odds)
    - log(p/1-p)
- the logistic function converts the log(odds) into probabilities
    - e^(log(odds))/(1+e^(log(odds)))

#### Ridge Regression:
- it is a technique part of regularization used to prevent overfitting
- it's called L2 norm
- uses the sum of squares of the magnitude of the coefficient 
- penalizes the coefficient of the linear model to a very smaller value at most but never the value is brought down to 0

- the ridge regression is very similar to the linear regression and effective when we have overfitting in linear regression algorithm
- in linear regession, we try to minimize the residual sum of squared
    - Decrease RSS
- whereas in ridge regression, we try to minimize the residual sum of squares + λ*slope^2
    - Decrease RSS + (λ*(slope^2))
        - λ => regularization parameter
            - λ ranges from 0 to +infinity
            - if λ value is large, then we might face high bias issues
- the role of lambda is to make the slope of the linear regression line less steeper than the OLS
    - as the λ value increase from 0 to + infinity the slope of the line will become even more less steeper
    - in other words, the λ value decreases the influence of independent feature on the output values

- when the line is steep, then the slope of the line is higher

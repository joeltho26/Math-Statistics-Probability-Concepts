#### Lasso Regression:
- it is part of the regularization technique to prevent overfitting
- it is called as L1 norm
- it penalizes the coefficient till the value are equal to 0, resulting in removing the unwanted feature for drawing the best fit line\
- the sum of absolute magnitude of the coefficients

- the lasso regression is very similar to the linear regression and effective when we have overfitting in linear regression algorithm
- in linear regession, we try to minimize the residual sum of squared
    - Decrease RSS
- whereas in lasso regression, we try to minimize the residual sum of squares + λ*slope
    - Decrease RSS + λ * slope
        - λ => regularization parameter
            - λ ranges from 0 to +infinity
            - if λ value is large, then we might face high bias issues
- the role of lambda is to make the slope of the linear regression line less steeper than the OLS
    - as the λ value increase from 0 to + infinity the slope of the line will become even more less steeper
    - in other words, the λ value decreases the influence of independent feature on the output values
- to choose the best value for λ, we perform cross validation
- as we increase the λ value, we clearly see the kink in the gradient descent curve

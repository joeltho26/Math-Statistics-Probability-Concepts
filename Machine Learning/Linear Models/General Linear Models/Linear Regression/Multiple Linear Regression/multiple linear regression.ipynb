#### Multiple Linear Regression:
- it is the same as linear regression, the only difference is that it has more than 1 input feature
- data points are present in more than 1 dimensions of data
- here we use adjusted R2 value to evaluate the goodness of fit
- Overfitting:
    - when the sample used for training and building the best fit line is small
- Undefitting:
    - when the data in non linear
    - when the model is not fit with the best fit
    - outliers present in the data

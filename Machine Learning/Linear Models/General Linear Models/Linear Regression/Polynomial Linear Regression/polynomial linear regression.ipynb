#### Polynomial Linear Regression:
- it is a variant from simple linear regression
- here it consist of 1 target variable and 1 independent variable
- used when the data is exponential in nature
- Overfitting:
    - when the sample used for training and building the best fit line is small
- Undefitting:
    - when the model is not fit with the best fit
    - outliers present in the data

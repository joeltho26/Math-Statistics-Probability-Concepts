#### Simple Linear Regression:
- used to perform prediction for regression problem statement
- represented by the equation of a straight line
- Overfitting:
    - when the sample used for training and building the best fit line is small
- Undefitting:
    - when the data in non linear
    - when the model is not fit with the best fit
    - outliers present in the data

- here in simple linear regression, we have one independent variables (feature) and 1 target variable used to perform prediction
- since the linear regression is denoted by equation of a straight line
    - the formula for a straight line is y = mx + c
        - y => dependent variable
        - m => slope
            - unit change in y (dependent variable) w.r.t to 1 unit change in x-axis (independent variable)
        - x => independent variable
        - c => intercept
            - the point at which the linear regression line that intercept the y-axis, when the independent value is 0

- we use the ordinary least squared method to find the best fit line for linear regression
    - Ordinary Least Square:
        - we calculate the sum of squared distance between the predicted line and the data points known as errors 
        - we make sure, we choose the best fitting line which has the lowest sum of square residuals calculated via cost function
            - cost function = 1/2 * Î£ (y-y^)^2

- to find the performance of the model, we use R squared
    - how well the independent variables explain variation of dependent variable
    - the goodness of fit
- the formula for R-squared
    - R2 = 1 - (RSS/TSS)
        - RSS => residual sum of squares i.e. the residuals from the fit line to the data points
        - TSS => residuals sum of squares from the mean of the target variable

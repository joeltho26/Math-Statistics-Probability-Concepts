#### ANOVA - Analysis of Variance:
- it is a variant of linear regression
- instead of continuous input data (simple linear regression), if input is categorical data to predict continuous output values
- imagine if the categorical variable is binary

Output:
- y (size) = m (slope, diet difference) * x (discrete input variable, high fat diet) + c (intercept, normal diet)
- the size prediction formula based on the line equation is 
    - Size = 1.5 + 0.7 * High Fat Diet
        - normal diet = 1.5 (intercept)
        - diet difference between High Fat vs Normal = 0.7 (slope) 

Model Performance:
- to evaluate the model performance, we need to use the RSS

Note:
- from binary/multi-label categorical variable, one category label is considered as a reference
    - eg: reference label is chosen with lowest average value (intercept), the other categorical labels will be multiplied with slope values
- difference in value between the 2 category i.e. the current category and the reference category will be acting as coefficient (slope)

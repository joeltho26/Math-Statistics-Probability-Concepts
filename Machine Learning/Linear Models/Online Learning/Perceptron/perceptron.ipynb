#### Perceptron:
- here the simple perceptron along with a linear activation function creates a linear line and used for linear regression problems
- simple perceptron consist of one summing function called neuron along with a activation function (linear) to perform regression
- the neurons are fired and not fired based on activation functions
- it also uses a optimizer to optimally increase the weights and bias 
- it also has a loss/cost function which uses gradient descent approach along with learning parameter to reduce the error/residual of the prediction, making the predictions more accurate

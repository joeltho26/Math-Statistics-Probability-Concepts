#### Passive Aggressive Algorithm:
- it is also called as online learning algorithm
- preferred for classification
    - online learning:
        - when we have one input data at a time to build, train the model for prediction
        - eg: Twitter
            - there is a large dataset for the model to train on certain use case. 
            - instead of using the whole dataset at once which will make the building and training process very slow, we use one data at a time to adjust the residuals 

Working:
- we first standardize/normalize the data first
- we make the weight value 'w' = 0
- we build a decision boundary which ranges between +1 and -1
- considering the prediction is binary 
    - the class 1 are on the side of +1 boundary
    - the class 0 are on the side of -1 boundary
- when we evaluate each input data and see if they adhere the constraints
    - constraints: 
        - +ve class data points should be i.e. d.w >= 1 
        - -ve class data points should be i.e. d.w <= -1 
            - d = new data points (new data)
            - w = the weight vector (decision boundary)

Aggressive:
- if any data that is misclassified will have the constraints not satisfied and the decision boundary is redrawn
    - the decision boundary is calculated via loss function updating the old weights
        - loss function: d.L.w
            - hinge loss
        - new weight:
            - Wnew = Wold + yLd
                - y = output label
                - L = loss function 
                - d = document

#### Assumptions of Linear Regression:
- Independence of errors
- No Multi-collinearity
- Exhibit Homoskedasticity
- Exhibit Linearity
- Exhibit Multi-variate normality of errors
- Absence of endogeneity

Independence of Errors:
- means that error of one value should not be related to the error of another value

No Multi-collinearity:
- means that no variables should be highly correlated with each other

Homoskedasticity:
- means that the variance of error (predictions) doesnt vary much with the independent variable change

Linearity:
- means that independent and dependent variables should exhibit linear relationship

Multivariate normality of errors:
- means that, the normality of errors i.e the errors exhibit a normal distribution of data

Absence of endogeneity:
- means that the absence of correlation between variables in a model and the error

#### Ordered Mean Target Guided:
- the category labels are encoded based on ```probability of output class lables```
- it will result in ```data leakage```
    - the output class information is shared into the encoded values of the categorical variable
    - to overcome data leakge:
        - ```Train-Test Split```:
            - perform encoding only on train dataset
            - map encoded values to test set 
        - ```K-Fold CV```:
            - Fold set: excluded from mean ordered encoding
            - K-1 set: used for mean ordered encoding
        - ```Regularization```:
            - creates a weighted average between category mean and global mean
            - ```Eᵢ = (Cᵢ * Mᵢ + λ * G)/(Cᵢ + λ)```
                - ```Cᵢ```: count of samples in category i
                - ```Mᵢ```: target mean for category i
                - ```G```: global mean
                - ```λ```: smoothing parameter
                    - ```↑``` value: pulls category mean closer to global mean
                    - ```↓``` value: moves category mean away from global mean

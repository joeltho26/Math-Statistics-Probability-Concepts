#### One Hot Encoding:
- it used to transform categorical features into numerical values
- for n feature it creates n-1 columns
- if we have more categories of feature, then it will be affected by curse of dimensionality
- we also remove last column to remove dummy variable trap

Before Encoding:

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/one_hot.png')
new_image = img.resize((250, 210))
display(new_image)
```

After Encoding:

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/one_hot2.png')
new_image = img.resize((600, 220))
display(new_image)
```

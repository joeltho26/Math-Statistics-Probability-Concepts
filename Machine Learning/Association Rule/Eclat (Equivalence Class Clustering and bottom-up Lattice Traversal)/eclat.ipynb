#### Eclat (Equivalence CLAss Transformation) Algorithm:
- it is used to identify the relationship and patterns in the itemset
- its a depth first search

Process of Eclat:
- Set min support
- take all the subsets from itemset which have support greater than the min support
- sort the subset by decreasing support 

- now lets consider the dataset below as the itsemset for apriori to find the relationship and patterns betwwen items in the itemset
- the min support count = 2
- support is the ratio of support count by total number of items in the itemset
    - Support Count/N
        - N => total number of items in the itemset

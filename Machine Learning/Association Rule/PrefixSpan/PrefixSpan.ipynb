#### PrefixSpan:
- used to identify the frequent subsequences from the sequence of itemsets

- we consider the itemset with their respective transaction as the data for PrefixSpan algorithm

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/data.png')
new_image = img.resize((290, 250))
display(new_image)
```

- at first, we set the min support = 2
- now calculate the occurence of each item in the itemset w.r.t transactions
    - a => 5
    - b => 5
    - c => 4
    - d => 1
    - e => 2
    - f =>1
- we also check what all items have satisified the min support = 2
    - a,b,c,e

- similar to H-Mine algorithm, we try to find the items following the a item:
    - the items following the prefix item is called projected databases i.e. prefix item = "a"
        - b,c,d,e
        - b,f
        - b,c,b,a
        - b,e,c
    - now within the projected database, we identify the frequent itemsets by satisfying the min support count
        - b,c,d,e,f i.e. unique items in the project database without the prefix "a"
            - b in the projected databases appear 5 times
                - b = 5
                - so (a,b) is a frequent item set
            - c = 3
                - so (a,c) is a frequent item set
            - d = 1
                - so (a,d) is not a frequent item set
            - e = 2
                - so (a,e) is a frequent item set
            - f = 1
                - so (a,f) is not a frequent item set
    - now the prefix item = (a,b) and d,f did not satisfy the min support so they are not considered
        - now the projected database for (a,b)
            - c,e
            - c,b,a
            - e,c
        - now within the projected database, we identify the frequent itemsets by satisfying the min support count
        - c,d,e i.e. unique items in the project database without the prefix "ab"
            - c = 3, in the projected databases appear 3 times
                - so (a,b,c) is a frequent item set
            - e = 2
                - so (a,b,e) is a frequent item set
    - now the prefix item = (a,b,c)
        - now the projected database for (a,b,c)
            - e
            - e
        - now within the projected database, we identify the frequent itemsets by satisfying the min support count
        - d,e i.e. unique items in the project database without the prefix "abc"
            - e = 2, in the projected databases appear 2 times
                - so (a,b,c,e) is a frequent item set
            

- likewise, we repeat the same for prefix b,c,d,e
- finally find the frequent items by their support count satisfying the min support count

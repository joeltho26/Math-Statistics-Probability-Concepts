- now we create the conditional database
    - calculating the conditional pattern base:
        - we look for the items with least support count
        - now looking at the tree, what are all the flows from root to reach the node I5 and their count
            - I2,I1 => 1
            - I2,I1,I3 => 1
        - I4 and their count
            - I2,I1 => 1
            - I2 => 1
        - I3 and their count
            - I2,I1 => 2
            - I2 => 2
            - I1 => 2
        - I1 and their count
            - I2 => 4
    - calculating the conditional FP tree:
        - we consider the items in the curly braces equal to the value written against it
        - we sum up the count of the items from all the curly braces and satisfy the min support count = 2
            - I5:
                - I2:2
                    - {I2,I1:1} => 1
                    - {I2,I1,I3:1} => 1
                - I1:2
                    - {I2,I1:1} => 1
                    - {I2,I1,I3:1} => 1
                - Therefore [I2:1, I1:2] as shown in conditional FP tree
            - I4:
                - I2:2
                    - {I2,I1:1} => 1
                    - {I2:1} => 1
                - Therefore [I2:1] as shown in conditional FP tree
            - I3:
                - I2:4
                    - {I2,I1:2} => 2
                    - {I2:2} => 2
                    - {I1:2} => 2
                - I1:2 (the value for I1 not equal to 4, since the I1=2 is received from left side and the I2=2 from the right side so we take them as separate entities)
                    - {I2,I1:2} => 2
                    - {I2:2} => 2
                    - {I1:2} => 2
                - Therefore [I2:4, I1:2], [I1:2] as shown in conditional FP tree
            - I1:
                - I2:4
                    - {I2:4} => 4
                - Therefore [I2:4] as shown in conditional FP tree

- similar to the apriori, we choose the itemset with the highest number of distinct items found in the frequent patterns generated columns
    - {I2,I1,I3:2} or {I2,I1,I5:2} => (l)
    - we choose I2,I1,I5 but generally, we calculate for both the sets above
- now take their non-empty subsets
    - I1
    - I2
    - I5
    - I1,I2
    - I2,I5
    - I1,I5
    - I1,I2,I5
- the output rule generated for non empty subset (s),
    - I1, output rule = I2,I5
    - I2, output rule = I1,I5
    - I5, output rule = I1,I2
    - I1,I2, output rule = I5
    - I2,I5, output rule = I1
    - I1,I5, output rule = I2
    - I1,I2,I5, l = null (neglect)

- the confidence formula can be also said as,
    - confidence = support count(l)/support count(s)
- we choose the rule that has confidence greater than min confidence
    - min confidence = 0.6
- now we calculate the confidence for the output rule generated for non empty subset,
    - I1, l = I2,I5, conf = 2/6 = 0.33
        - I1 = 6
        - I1,I2,I5(l) = 2
    - I2, l = I1,I5, conf = 2/7 = 0.30
    - I5, l = I1,I2, conf = 2/2 = 1
    - I1,I2, l = I5, conf = 2/4 = 0.5
    - I2,I5, l = I1, conf = 2/2 = 1
    - I1,I5, l = I2, conf = 2/2 = 1 

- now considering the min confidence, the below rules satisfy strong association
    - I1,I2

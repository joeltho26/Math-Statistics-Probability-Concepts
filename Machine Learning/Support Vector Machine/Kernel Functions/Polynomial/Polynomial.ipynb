#### Polynomial Kernel:
- calculate the relationship between each data points by tranforming data points imaginarily
    - Classification: making data points linearly separable
    - Regression: making data points follow linearity property
- ```polynomial property```: 6(x)^3 + 3(x)^2 + 5x + 2 = 0
- polynomial kernel will transform the data distribution into a polynomial distribution i.e. similar to exponential distribution
- consist of 2 hyperparameter ```'r'``` & ```'d'```
    - ```r``` = coefficient
        - choose variation of fractional or whole number values
    - ```d``` = degree of polynomial
        - choose variation of whole number values
- taking the ```dot product rule``` by substituiting values for 'd' & 'r', we will see the relationship of data points in higher dimension

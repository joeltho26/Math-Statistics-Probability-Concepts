#### Polynomial Kernel:
- used to not move the data points from a lower dimension to higher dimension
- used to calculate the relationship between each data points and choose the support vector classifier
- uses the kernel trick
    - calculates the relationship between data points without converting the data points into higher dimensions

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/poly.png')
new_image = img.resize((210, 60))
display(new_image)
```

- the value of d (degree of the polynomial) is chosen based on the cross validation
- a, b are the 2 data points
- r is the coefficient (manually chosen)

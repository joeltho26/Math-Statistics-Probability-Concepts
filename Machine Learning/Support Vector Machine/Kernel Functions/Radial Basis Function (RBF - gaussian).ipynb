#### Radial Basis Function:
- used to not move the data points from a lower dimension to higher dimension
- used to calculate the relationship between each data points and choose the support vector classifier in infinite dimensions
- uses the kernel trick
    - calculates the relationship between data points without transforming the data points into higher dimensions
- performs like a Weighted Nearest Neighbor model
    - data points nearer will have more influence than the data points far away from it

- the formula for radial basis is shown below
- since the radial basis works with dimensions in infinite dimesions, its diffcult to visualize it

consider the example:
- Î³ = 1/2

```Drawing the marginal line/hyperplane```:
- we draw a line joining the the identified support vectors
    - ```support vectors```: 2 data points that belong to opposite class and closer to one another
- now, we take the average of distance between the 2 points
- draw a perpendicular line to the initial line that joined the 2 support vectors and goes through mean
- the perpendicular line is called the ```marginal line/hyperplane``` based on the number of dimensions i.e. the features considered
- the line/hyperplane helps to splits the classes in binary/multi-class classification

```Marginal Distance```:
- the distance between 2 support vectors is called ```marginal distance```

```Optimization factor``` - maximize the marginal distance:
- the marginal distance should be ```maximum``` to attain good model with best prediction
    - i.e. both sides of the hyperplane should be fairly large
- to optimize the margin distance value, we take the differential equation of both +ve and -ve hyperplanes
- by adding a normalized |w| in the denominator on both sides we removed the W component

```Hard Margin```:
- when there is an outlier, the marginal hyperplane will be redrawn to satisfy each and every data point
- this fails to capture the underlying data pattern and unable to build a generalized model
- practical datasets have outliers and hard margin is not feasible approach

```Optimization Factor for Hard Margin```:
- ```Minimize (|w|)^2/2```

```Prediction for Hard Margin```:
- the output value will be a numerical value instead of a label
- the numerical value will be compared with the threshold value of category bucket
- which ever range it falls into, that output label will be chosen
- ```f(x) = W.Xnew + b```

```Soft Margin```:
- preferred choice when outliers exist in the dataset
- it allows for misclassified data points in the dataset and marginal hyperplane is not redrawn
- generalization of the model capturing the underlying data pattern help to build a better model with accurate prediction
- the number of misclassified data points is controlled by a hyperparameter C
- the misclassified data points are slack variables which are used to calculate the errors in terms of hinge loss and by penalizing minimize it

```Optimization factor for Soft Margin```:
- ```C``` = number of errors/misclassified data points
    - C is the cost parameter which is inversely related regularization
    - value of C should neither be small or large, bias-variance trade off
- `````` = value of the errors
    - also called as slack variables
    - represented by hinge loss

```Prediction for Soft Margin```:
- the output value will be a numerical value instead of a label
- the numerical value will be compared with the threshold value of category bucket
- which ever range it falls into, that output label will be chosen
- ```f(x) = W.Xnew + b```

```Optimization of SVM```:

- ```Shrinking```:
    - the non-support vectors are temporarily made inactive during the training purpose i.e. creating the hyperplane and added back later

- ```tol (Tolerance)```:
    - stopping criteria of the optimization algorithm used to train the SVM
    -  tol:  precision,  accuracy,  training time.
    -  tol:  precision,  accuracy,  training time.
    - CV used to find the best value

- ```max_iter```:
    - a parameter that sets an upper bound on the number of iterations to optimize the algorithm

- ```decision_function_shape```:
    - used for multiple class prediction
        - ovo = one vs one
        - ovr = one vs rest

- ```break_ties```:
    - If true, decision_function_shape='ovr', and number of classes > 2, predict will break ties w.r.t confidence values of decision_function; otherwise the first class among the tied classes is returned. 
    - breaking ties comes with  computational cost

#### Linear SVM:
- the data points are linearly separable
- we build an 'ε' epsilon in-sensitive tube
ensure most of the points are within the epsilon tube

```Choose value of Epsilon 'ε'```:
- prior knowledge of the data
- cross validation (CV)
- note:
    - the 'ε' value is chosen before training

```Hard margin```:
- no data point is allowed to lie outside of the epsilon tube
- when outliers are present, the initially fit epsilon tube is re-adjusted to satisfy the outlier
- if outliers are extreme, the model becomes impossible to fit a smooth regression function
- for practical datasets, SVR Hard Margin is not recommended

```Optimization Factor for Hard Margin```:
- ```Minimize (|w|)^2/2```

```Prediction for Hard Margin```:
- ```f(x) = W.Xnew + b```

```Soft Margin```:
- preferred choice when outliers exist in the dataset
- it allows for outlier data points to be present in the dataset and epsilon insensitive tube is not re-adjusted
- generalization of the model capturing the underlying data pattern help to build a better model with accurate prediction
- the number of outlier data points is controlled by a hyperparameter C
- the outlier data points are slack variables which are used to calculate the errors in terms of hinge loss by penalizing and minimize it

```Optimization factor for Soft Margin```:
- ```C``` = number of errors/outlier data points
    - C is the cost parameter which is inversely related regularization
    - value of C should neither be small or large, bias-variance trade off
- ```ε, ε*``` = value of the errors
    - also called as slack variables
    - represented by hinge loss
    - ```ε``` = outlier data points that lie outside the upper bound of the epsilon tube
    - ```ε*``` = outlier data points that lie outside the lower bound of the epsilon tube
    - called the L1 norm

```Prediction for Soft Margin```:
- ```f(x) = W.Xnew + b```

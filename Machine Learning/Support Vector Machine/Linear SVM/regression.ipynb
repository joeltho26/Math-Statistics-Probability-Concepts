#### Linear SVM:
- the data points are linearly separable
- for regression, when we build the hyperplane and its epsilon tube, ensure most of the points are within the epsilon tube

- at first, we assign epsilon a value for the width of the epsilon tube i.e. marginal distance from the hyperplane
    - if we have prior knowledge of the data, we can assign the value directly
    - if we do not have idea, then choose the value based on cross validation

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/svr1.png')
new_image = img.resize((560, 370))
display(new_image)
```

- ζi = are the slack data points/variables for regression i.e. data that lie outside of the tube above the hyperplane
- ζ*i = are the slack data points/variables for regression i.e. data that lie outside of the tube below the hyperplane

Note:
- generally slack variables are used in classification as well i.e. to call misclassified data points
- we will penalize the slack variables

- for hard margin as example below,
    - we do not allow for any data point to be outside of the epsilon tube or tolerance level
    - look at the constraints shown below

- but most times, data comes with outliers and are not perfect, so as a result we will find data points even outside the epsilon tube which are called as Zeta (top and bottom)
- so to allow zeta variables in SVR and to proceed, with soft margin as shown below
    - C = Regularization parameter
    - ζ, ζ* = slack variables or the support vectors

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/svr4.png')
new_image = img.resize((560, 320))
display(new_image)
```

- when C value high = Variance increase & Bias decrease
    - leads to overfitting
- when C value low = Variance decrease & Bias increase
    - leads to underfitting
- C parameter is the inverse of regularization
    - regularization = is forcing the model to be strict and not allow any errors and penalizing heavily
- (ξi + ξi∗) = L1 norm

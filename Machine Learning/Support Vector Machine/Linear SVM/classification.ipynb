Optimization of SVM:
- Shrinking:
    - the non-support vectors are temporarily made inactive during the training purpose i.e. creating the hyperplane and added back later
- tol (Tolerance):
    - stopping criteria of the optimization algorithm used to train the SVM
    - Small tol: Higher precision, potentially better accuracy, longer training time.
    - Large tol: Lower precision, potentially slightly lower accuracy, faster training time.
    - Also use CV to find the best value for it
- max_iter:
    -  is a parameter that sets an upper bound on the number of iterations the optimization algorithm  will perform during the training process. 
    - It acts as a safeguard to prevent the optimization from running indefinitely if it's not converging quickly enough or if there are issues with the data or parameters.
- decision_function_shape:
    - used for multiple class prediction
        - ovo = one vs one
        - ovr = one vs rest
- break_ties:
    - If true, decision_function_shape='ovr', and number of classes > 2, predict will break ties according to the confidence values of decision_function; otherwise the first class among the tied classes is returned. 
    - note that breaking ties comes at a relatively high computational cost compared to a simple predict.

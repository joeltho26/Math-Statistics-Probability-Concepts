#### Non-Linear SVM:
- used when data points are non linearly separable
- we implement kernels to transform and make data points linearly separable

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/nsvm.png')
new_image = img.resize((490, 40))
display(new_image)
````

```Kernels```:
- ```transform``` the data point into a ```higher dimension```
- data points become linearly seprable when transformed from original dimension to a higher one 
- evaluate the ```relationships``` between each and every data point
- based on relationships between data points in higher dimension, the support vector classifier produces a ```marginal hyperplane```

```Kernel Trick```:
- special ability of kernels 
- evaluate the relationships between data points in higher dimensions ```without transforming``` the data point into higher dimension

```Types of Kernels```:
- Polynomial
- Sigmoid 
- Radial Basis Function
- Linear

```Prediction with Kernels```:
- the output value will be a numerical value instead of a label
- the numerical value will be compared with the threshold value of category bucket
- which ever range it falls into, that output label will be chosen
- ```f(x) = Σ αi * Yi * K(Xnew,Xi) + b```
    - ```αi/Wi``` = weight
    - ```Yi``` = observed label
    - ```Xnew``` = new input data point
    - ```Xi``` = existing input data point
    - ```b``` = bias

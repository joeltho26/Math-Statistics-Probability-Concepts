#### Non Linear SVR:
- used when data points are not linear
- we implement kernels to transform and make data points follow linearity

Kernels:
- ```transform``` the data point into a ```higher dimension```
- data points follow ```linearity``` when transformed from original dimension to a higher one 
- evaluate the ```relationships``` between each and every data point
- based on relationships between data points in higher dimension, the support vector regressor produces a ```epsilon tube``` that fits the data by penalizing and minimizing the ```hinge loss``` score

```Kernel Trick```:
- special ability of kernels 
- evaluate the relationships between data points in higher dimensions ```without transforming``` the data point into higher dimension

```Types of Kernels```:
- Polynomial
- Sigmoid 
- Radial Basis Function
- Linear

```Prediction with Kernels```:
- ```f(x) = Σ αi * K(Xnew,Xi) + b```
    - ```αi/Wi``` = weight
    - ```Xnew``` = new input data point
    - ```Xi``` = existing input data point
    - ```b``` = bias

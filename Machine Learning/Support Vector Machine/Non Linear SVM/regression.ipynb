#### Non Linear SVR:
- to perform support vector classifier on non linear data
    - we need to increase the dimensions of data 
    - calculate the relationship between each datapoints 
    - then choose the support vector classifier to separate the classes
        - the support vector classifier is evaluated as the same as linear SVM

- to increase the dimensions and choose the support vectors classifier, we need kernels
- Kernels
    - Polynomial
    - Sigmoid 
    - Radial Basis Function
    - Linear

- now consider, we have feature with binary output values in a 3d space
- to make the predictions for Linear SVM soft classifier shown below
    - Î±i = weights of the data points
    - b = bias
    - K(xi . x) = kernel function which measures similarity between new data point multiplied with support vector
        - x = new data point 
        - xi = support vectorss

- now lets consider an example with prediction of SVR for non linear use case:
- support vectors:
    - x_1 = [1.5, 2]
    - x_2 = [2.0, 1]
    - x_3 = [2.5, 3]
- Lagrange Multipliers (Î±_i - Î±_i*) also known as weights
    - Î±_1 - Î±_1* = 0.2
    - Î±_2 - Î±_2* = -0.5
    - Î±_3 - Î±_3* = 0.3
- Bias (b): b = 1.0
- RBF Kernel Parameter (Î³): Î³ = 0.5
- choosing RBF kernel
    - K(x, x_i) = exp(-Î³ * ||x - x_i||^2)
        - x is the new input data point.
        - x_i is a support vector.
        - ||x - x_i||^2 is the squared Euclidean distance between x and x_i.
        - Î³ is the kernel parameter.

#### Multi-class in SVM:
- we have two strategies to perform multiclass in SVM
    - ovo = one vs one
    - ovr = one vs rest

##### OVR (One vs One):
- comparing two output labels at a time and ignoring the rest

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/ovo.png')
new_image = img.resize((300, 300))
display(new_image)
```

- now consider, we have test data point

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/ovo_test.png')
new_image = img.resize((300, 300))
display(new_image)
```

- now we find the new datapoint closer to which output label and repeat for all the combination.
- choose the output class for the new data point based on the majority of votes
    - we see purple has the majority of votes, so the output label for the new data point is purple

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/ovo_vote.png')
new_image = img.resize((300, 120))
display(new_image)
```

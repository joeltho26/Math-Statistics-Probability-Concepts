#### Multi-class in SVM:
- one of the strategies to perform multi-class classification in SVM
    - OVR = ```one vs rest```
- preferred when the number of output labels > 10
- faster, but less accurate in performance compared to OVO

##### OVR (One vs Rest):
- eg: 3 output classes
    - there will be 3 separate classifier built and trained
        - Classifier 1: Blue vs (Red, Green)
        - Classifier 1: Green vs (Red, Blue)
        - Classifier 1: Red vs (Blue, Green)
    - final output chosen based on the highest +ve value from the classifier i.e. ```distance``` value

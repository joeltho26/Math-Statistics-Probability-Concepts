#### Multi-class in SVM:
- one of the strategies to perform multi-class classification in SVM
    - OVO = ```one vs one```
- preferred when the number of output labels <= 10
- more accurate, but slower compared to OVR when output labels >= 10

##### OVR (One vs One):
- compare only two class data points at a time and ignore the rest
- there are 3 classifiers being built and trained
    - Classifier 1: Blue vs Red
    - Classifier 1: Blue vs Green
    - Classifier 1: Green vs Red
- the final output label is chosen based on the majority of votes

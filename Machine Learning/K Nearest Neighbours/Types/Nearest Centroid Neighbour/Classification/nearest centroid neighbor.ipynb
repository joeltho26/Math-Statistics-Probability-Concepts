#### Nearest Centroid Neighbor:
- used to perform ```only classification``` problem
- it very similar to regular KNN algorithm
- it does not work on K or R value. Instead, a ```centroid``` is calculated for each class based on their data points
- a ```new test/query point``` is classified based on the ```nearest``` centroid of a class from the query point

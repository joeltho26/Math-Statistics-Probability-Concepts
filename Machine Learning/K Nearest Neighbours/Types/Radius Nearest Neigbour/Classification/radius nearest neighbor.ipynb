#### Radius Nearest Neighbor
- instead of choosing the optimal K nearest neighbor value, the ```radius``` value for the circular region is chosen

```Process of Radius Nearest Neigbors```:
- first, ```project the data points``` onto the ```N dimensional space``` (N=```number of features```) with output labels
    - scalar plot
- consider, a ```query/new test``` data point to be classified
- start with choosing the ```R (radius)``` value i.e. the radius of the circular region which encompasses the number of data points
    - using hyperparameter tuning and CV 
- choosing R value leads to overfitting and underfitting issues
    - very ```↓ R``` value (which encompasses only 1 data point), overfitting
        - each data point will influence in the prediction of output label, ↑ variance
    - very ```↑ R``` value, underfitting
        - when the dataset is ```imbalanced```
        - i.e. ```95``` records belong to class A and ```5``` records belong to class B
        - the class B will always be the output chosen 
    - for specific R values,    
        - if the number of data points associated with each class labels is the ```same or even``` inside the Radius
        - then the output labels will be based on

            - ```nearest data point class label```
        
            - ```random selection```

- after choosing R value, calculate the nearest neighbor points to query point using the ```distance metrics```
- calculating the nearest neighbor points,
    - i.e. the nearest neighbour data points are identified as the data point's ```distance``` from query point ```< R``` (radius value)
- after identifying the nearest neighbor points, look at the data points with ```majority``` of output labels associated which will be the final output label
- Eg: from the below image with ```R = 1```, the final output label is class Blue based on the majority of data points belonging to class Blue which were considered as the nearest neighbors

#### K Nearest Neighbors:
- KNN solves regression problems

```Process of K-Nearest Neigbors```:
- first, ```project the data points``` onto the ```N dimensional space``` (N=```number of features```) with output regression values
    - scalar plot
- consider, a ```query/new test``` data point for which the regression value needs to be calculated
- start with choosing the ```K``` value i.e. the number of neighbours to be considered to perform prediction
    - using hyperparameter tuning and CV 
- choosing K value leads to overfitting and underfitting issues
    - ```K = 1```, overfitting
        - each data point will influence in the prediction of output value, ↑ variance
    - ```K``` very ```↑```, underfitting
- after choosing K value, calculate the K nearest neighbor points to query point using the ```distance metrics```
- after identifying the nearest neighbor points, take the ```average``` of all continuous output values associated with it
- Eg: from the below image with K = 3, the final output value will be the average of all the 3 data points considered as nearest neighbour

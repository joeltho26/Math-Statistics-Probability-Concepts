#### Weighted K Nearest Neighbors:
- solve classification problems
- works very similar to simple KNN algorithm
- ```Weighted K-Nearest Neighbors```: 
    - ```data points closer``` to query data point has ```more say/influence/weight``` than the ```data points farther``` from the query data point
    - ```Weight = 1/Distance```
- the ```prediction of output labels``` depends on the ```weight values``` for each class data points considered as nearest neighbour

```Example Working of Weighted KNN```:
- consider binary classification dataset shown below
- number of Nearest Neighbor, K = 5 chosen
- calculate the K nearest neighbors, based on the nearest distances calculated between the query point and existing data points with output labels
- we assign weight values along with the distances calculated for each data points closer to the query point
    - ```weight = 1/distance```
- ```Prediction```:
    - Weights:

        - ```Red = 5 + 2 = 7```

        - ```Green = 1.4 + 0.8 + 0.6 = 2.8```
        
- we assign the query points with the output label, based on the highest sum of weights for a class

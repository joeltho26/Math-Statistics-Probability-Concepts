#### K Nearest Neighbors:
- solves both classification and regression problems

##### Process of K-Nearest Neigbors:
- Aft first, we have the records in a dataset with output label available
- Now, consider if we have to predict the output label of a new data point in the scaler plot shown below

```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/knn_eg.png')
new_image = img.resize((480, 450))
display(new_image)
```

- we need to first assign the K value manually or by cross validation and hyper parameter tuning
- if K values = 3, then we would consider 3 data points closer to the new data point and calculate the distance between them using Euclidean, Manhattan, or any distance formula.
- after finding the closest data points to the new query data point and see the majority of the labels and we assign the query data point with the majority labels
- if K values is even and the majority of votes are the same for both the classes
    - random selection
    - prioritize by distance
        - class closest to the query point
- based on the K-values, we can overfit or underfit result
    - if K value = 1 or if the k value is small
        - will result in high variance and overfit
    - if the k value is large and when data is imbalanced
        - will result in high bias and underfit

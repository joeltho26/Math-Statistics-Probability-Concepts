#### KD Trees:
- used to find the K-nearest neighbours in multiple dimensions
- performs better if the data points are between lower and moderate level of dimensions and when data is normally distributed i.e. evenly
- the K in KD Tree is denoted as the properties of a node i.e. the number of dimensions
    - (5,4) => 2 properties i.e. 2D
    - (5,8,2) => 3 properties i.e. 3D
- works on the principal of Binary Search Tree
    - value lesser than the parent will go to the left
    - value greater than the parent will go to the right

##### Process of KD Trees:
- at first, consider we have few data points for which we will create the kd tree
    - (5,4), (2,6), (13,3), (8,7), (3,1), (10,2)
- we sort the data point based on the x-axis dimension and take the median value
    - (2,6), (3,1), (5,4), (8,7), (10,2), (13,3)
- now the median value based on the x-axis will be the root node
- since the number of data points are even, median is not a single datapoint, therefore we can choose
    - left side of median
    - right side of median
    - center of the median, not a data point present already
- the dimensions used for splitting the node are used alternatively

- the first split from (5,4) root node was based on x-axis so the split is perform based on x-axis
    - Root = (5,4)
    - Internal node left: left most value (2,6)
    - Internal node right: right most value (13,3)
- the next split is performed based on the y-axis for (2,6) & (13,3)
    - Left - Parent = (2,6)
    - Internal node left: left most value (3,1)
    - Right - Parent = (13,3)
    - Internal node left: left most value (10,2)
    - Internal node right: right most value (8,7)

- the smallest distance from the query point 
    - we have the distance value for [9,4] the lowest for (8,7) rectangular strip
- now we check for the backtracking condition
    - we draw a hypersphere by keeping the data point [9,4] as the center and the radius as the distance between ((9,4) & (8,7)) and watch out if there are any split lines are being intersected
    - if split lines from the constructed tree are intersect then, we perform backtracking on the nodes for which we did not calculate the minimum distance 
- backtracking is done until the condition is failed

#### Ball Trees:
- used to find the ```nearest neighbours``` in KNN
- performs better if the data points are in ```higher dimensions``` and ```sparsely distributed```

##### Process of Ball Tree:
- ```project all data points``` onto the ````N dimensional feature space``` and draw a ```single large hypersphere``` that encompasses all datapoints within it

```Building Hypersheres```:

```Confirm the actual nearest neighbor```:
- take the ```radius``` value i.e. ```distance from the nearest neighbor data point and the query point```

- if other hyperspheres intersects, repeat the same step of calculating the distance of each data point within the hypersphere and find the new nearest neighbour if the distance calculated is less than the previous nearest neighbor

#### Cosine Similarity & Distance:

```Cosine Similarity```:

- used to calculate the ```dissimilarities between 2 string``` such as words and are measured in terms of ```angle differences```
- focuses on the ```direction``` rather than ```magnitude```

```Similarities based on angle 'θ'```:

- ```0 <= θ < 90```: positive similarity

- ```θ = 90```: no similarity

- ```90 < θ <= 180```: negative similarity

```Distance Measure w.r.t Cosine Similarity```:

- ```Cosine Distance = 1 - cosine similarity```

```Soft Cosine Measure (SCM)```: 
- an extension of cosine similarity that accounts for the ```semantic``` and ```contextual similarity``` of features
- measures the ```similarity between two vectors``` by considering the ```similarity between their features``` i.e. words.
- Unlike standard cosine similarity, which treats all ```features``` as ```independent```, SCM accounts for semantic relationships

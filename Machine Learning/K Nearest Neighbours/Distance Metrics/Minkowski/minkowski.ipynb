#### Minkowski Distance:
- ```generalization``` of manhattan and euclidean distances
- ```p``` value decides the type of distance calculation
    - ```p = 1```, Manhattan 

    - ```p = 2```, Euclidean

    - ```1 < p < 2```, Minkowski

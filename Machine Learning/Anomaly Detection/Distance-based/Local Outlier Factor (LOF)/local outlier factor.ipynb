#### Local Outlier Factor:
- used to detect anomaly or outlier
- used to identify local outliers

- O3 => global outlier
    - use distance based metrics
- O1 => local outlier
    - use density based algorithms
- we have a LOF score to determine whether a point is an outlier or normal point

- consider we choose the data point p and we calculate the K distance
    - the K distance is the distance from the p point to K-th nearest neighbor
    - this distance defines the radius around the point p

- calculate the reachability distance 
    - the reachability distance for all the Kth Neigbour points will be the K-th distance value between point p and Kth nearest neighbour 
    - now, while calculating the reachability distance between point "p" and the point "o" outside the K-th neigbours, then we take the max distance between them.
        - reachability_distance(p, o) = max(k-distance(o), distance(p, o))

- calculate the local reachability density of the point "p"
    - local reachability density is the inverse of average reachability distance of p to it k nearest neighbors
    - if reachability distance is small, LRD is high
    - if reachability distance is large, LRD is low

- calculate the LOF score of the point using local reachability density of its neighbors and the point itself
    - if LOF = 1, likely an inliner
    - if LOF < 1, is an inliner and also maybe a core point
    - if LOF > 1, likely an outlier
- the LOF compares the density of the K-th neighbour from point p and the K-th neighbor from K-th neigbour of point p
    - if their density is similar and if they are similar we get the LOF score equal to 1 or lesser than else it will be greater than 1

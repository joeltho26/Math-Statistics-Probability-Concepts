#### Isolation Forest:
- it an ML algorithm used to identify anomalies data in the dataset
- it similar as Extra Trees algorithm
    - uses pasting instead of bootstrapping
        - row sampling without replacement i.e. the random subset of data given to the models (connected in parallel) to train without any overlap with other subsets
    - splitting of the node while building the tree is random
- since anomalies are rare and distinct, they are identified in Isolation forest by evaluating the path length (i.e. the number of edges from root to leaf node)


```python
from PIL import Image 
from IPython.display import display

img = Image.open('images/isolation2.png')
new_image = img.resize((800, 300))
display(new_image)
```

- if the path length is long then its a normal data point and if the path length is less then they are denoted as anomalies
- we build a specific number of tree and we take the average path length 
- the average path length for all isolation trees in the forest derives the anomaly score
    - anomaly score ranges 0-1
- lower average path length results in higher anomaly score i.e. its an anomaly point
- higher average path length results in lower anomaly score i.e. its an normal point

    - m = no. of data points

- In scikit learn, decision_function returns values where lower means anomalous
- contamination hyper parameter is used as a threshold for anomaly score to mark a data point as anomaly or normal data point and the values for contamination parameter ranges between 0-0.5
